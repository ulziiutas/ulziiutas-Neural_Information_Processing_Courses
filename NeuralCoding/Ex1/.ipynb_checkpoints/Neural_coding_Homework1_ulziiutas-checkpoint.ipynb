```python
from IPython.display import display, Math, Latex

# Libraries you might need 
from matplotlib import pyplot as plt
%matplotlib inline
from scipy.stats import bernoulli

import numpy as np
import statistics
import time
from numba import jit, cuda, vectorize
```

```python
# You can use this function for plotting raster plots (spikes), or you can use your own defined function
def raster(event_times_list, color='k', t_max=10.):
    """
    Creates a raster plot
    Parameters
    ----------
    event_times_list : iterable
                       a list of event time iterables
    color : string
            color of vlines
    Returns
    -------
    ax : an axis containing the raster plot
    https://scimusing.wordpress.com/2013/05/06/making-raster-plots-in-python-with-matplotlib/
    """
    plt.figure(figsize=(20,3))
    ax = plt.gca()
    for ith, trial in enumerate(event_times_list):
        plt.vlines(trial, ith + .5, ith + 1.5, color=color)
    plt.ylim(.5, len(event_times_list) + .5)
    ax.set_xlim((0,10))
    return ax
```

1) Take interval [0,10 sec]. We will simulate spikes within this interval to form a Poisson spike train. To do so we are going to use exponenetial distribution for inter-spike intervals. First spike-time (relative to time 0) is drawn from the exponential distribution with parameter $\tau$ , thus $t_1 \sim Exp(\tau) = 1/\tau e^{-t_1/tau}$. All subsequent spikes are determined relative to the previous via $t_{k+1} = t_k + \Delta t$ with $\Delta t \sim Exp(\tau)$. Stop when the new spike-time exceed 10 sec. (1.5 points per each sub-task, tasks with a "*" give additional points)

a) Plot one spike train for each of the following parameter values: $\tau = 1$sec, $\tau = 0.1$sec, $\tau = 0.01$sec  

```python
@jit(nopython = True)
def spike_gen(tau):
    max_time = 10;
    spike_train = [];
    for i in range(len(tau)):        
        deltaT = np.random.exponential(tau[i], int(15/tau[i]));
        last_value = deltaT[0];
        j = 1;
        tmp = [];
        while j < len(deltaT) and last_value <= max_time:
            tmp.append(last_value);
            last_value = last_value + deltaT[j];
            j += 1;            
        spike_train.append(np.array(tmp));        
    return spike_train
```

```python
tau = np.array([1, 0.1, 0.01]);   

start = time.time()
spike_train = spike_gen(tau);
end = time.time()
print(end - start)
raster(spike_train);
```

b) Simulate 100.000 spike trains for each parameter in a) and compute the total number of spikes (i.e. the “spike count”) for each. Plot a histogram over the spike counts for each of the three conditions, as well as the mean and the variance.

```python
@jit(nopython = True)
def spike_gen_count(tau):
    max_time = 10;
    spike_train_count = [];
    for i in range(len(tau)):        
        deltaT = np.random.exponential(tau[i], int(30/tau[i]));        
        last_value = deltaT[0];
        j = 1;
        tmp = [];
        while j < len(deltaT) and last_value <= max_time:
            tmp.append(last_value);
            last_value = last_value + deltaT[j];
            j += 1;
        if last_value < max_time:
            print('failed');
        spike_train_count.append(len(tmp));      
    return spike_train_count
```

```python
N = 100000;
spike_train_count = [];

start = time.time();
for j in range(N):
    tmp = spike_gen_count(tau);
    spike_train_count.append(tmp);
spike_train_count = np.array(spike_train_count);
spike_train_count = spike_train_count.transpose();
end = time.time()
print(end - start)
```

```python
plt.figure(figsize=(15, 5));
plt.suptitle('Histogram of Spike Count for different tau', fontsize=16);

for i in range(len(tau)):        
    plt.subplot(1,3,i+1);
    plt.hist(spike_train_count[i], bins=20);
    plt.title('Tau = {}'.format(tau[i]));    
    mean = np.mean(spike_train_count[i]);
    var = np.var(spike_train_count[i]);
    
    print('Mean of Tau {} = {}'.format(tau[i], mean));
    print('Variance of Tau {} = {}\n'.format(tau[i], var));
```

c) Draw 10.000 samples $x_k \in \{0, 1\}$ independently from a Beronulli distribution $P (x_k ) = \lambda^{x_k} (1 − \lambda)^{1−x_k}$ with $k = 1, \ldots , 10.000$ for the following three cases: $\lambda = 0.1$, $\lambda = 0.01$, $\lambda = 0.001$. Plot the random vector $x$ as a spike train for each of the three conditions.

```python
# @jit(nopython = True)
def spike_gen_ber(lambda_arr):    
    N = 10000;
    spike_train = [];
    
    for i in range(len(lambda_arr)):        
        
        tmp = np.array(bernoulli.rvs(size=N,p=lambda_arr[i]));
        boolArr = (tmp == 1);
        indeces_arr = np.asarray(np.where(boolArr));
        
        indeces_arr = indeces_arr/1000;
        spike_train.append(indeces_arr);
        
    return spike_train
```

```python
lambda_arr = np.array([0.1, 0.01, 0.001]);
spike_train_ber = [];

start = time.time();

spike_train_ber = spike_gen_ber(lambda_arr);
    
end = time.time()
print(end - start)

raster(spike_train_ber);
```

d) Simulate 100.000 spike trains for each parameter in c) and compute the total number of spikes (i.e. the “spike count”) for each. Plot a histogram over the spike counts for each of the three conditions, as well as the mean and the variance. How do the results differ from those b)?

```python
@jit(target = 'gpu')
def spike_gen_ber_count(lambda_arr):
    N = 10000;
    spike_train_count = [];
    
    for i in range(len(lambda_arr)):
        
        tmp = np.random.binomial(N, lambda_arr[i]);
        spike_train_count.append(tmp);
    
    return spike_train_count
```

```python
N = 100000;
spike_train_ber_count = [];

start = time.time();

for j in range(N):
    tmp = spike_gen_ber_count(lambda_arr);
    spike_train_ber_count.append(tmp);
    
spike_train_ber_count = np.array(spike_train_ber_count);
spike_train_ber_count = spike_train_ber_count.transpose();

end = time.time()
print(end - start)
```

```python
plt.figure(figsize=(15, 5));
plt.suptitle('Histogram of Spike Count for different Lambda', fontsize=16);

for i in range(len(lambda_arr)):        
    plt.subplot(1,3,i+1);
    plt.hist(spike_train_ber_count[i], bins=20);
    plt.title('Lambda = {}'.format(lambda_arr[i]));    
    mean = np.mean(spike_train_ber_count[i]);
    var = np.var(spike_train_ber_count[i]);
    
    print('Mean of Lambda {} = {}'.format(lambda_arr[i], mean));
    print('Variance of Lambda {} = {}\n'.format(lambda_arr[i], var));
```
