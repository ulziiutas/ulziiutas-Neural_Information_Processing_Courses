## E2.b

```python
s2 = np.arange(1.01,10,0.01);
mde = np.zeros(len(s2));

for i in range(len(s2)):    
    mde[i] = min(0.25, 0.5*(s2[i]**2/(1+s2[i]**2))) + min(0.25, 0.5*(1/(1+s2[i]**2)));
```

```python
plt.figure(figsize=(10,3));
plt.plot(s2, mde);
plt.title('MDE as function of s2');
plt.ylabel('MDE');
plt.xlabel('s2');
```

## E2.c

```python
I_js = np.zeros(len(s2));

for i in range(len(s2)):    
    I_js[i] = -0.5 * (-1 - s2[i]**2 * np.log2(s2[i]**2) / (1 + s2[i]**2) + np.log2(1+s2[i]**2));
```

```python
plt.figure(figsize=(10,3));
plt.plot(s2, I_js);
plt.title('Jensen-Shannon Divergence as function of s2');
plt.ylabel('Ijs');
plt.xlabel('s2');
```

```python
def h(x):
    return -(1-x) * np.log2(1 - x) - x * np.log2(x);
```

```python
u_bound = (1/2 - 1/2*I_js);
l_bound = np.zeros(len(s2));

for i in range(len(s2)):   
    
    sub = 1 - I_js[i];    
    h_b = lambda p: h(p) - sub;
    l_bound[i] = scipy.optimize.newton_krylov(h_b, xin=0.25);
    
```

```python
plt.figure(figsize=(10,3));
plt.plot(s2, u_bound, label = 'Upper Bound');
plt.plot(s2, mde, label = 'MDE');
plt.plot(s2, l_bound, label = 'Lower Bound');
plt.title('Jensen-Shannon information Upper bound as function of s2');
plt.ylabel('Ijs');
plt.xlabel('s2');
plt.legend();
plt.show();
```
