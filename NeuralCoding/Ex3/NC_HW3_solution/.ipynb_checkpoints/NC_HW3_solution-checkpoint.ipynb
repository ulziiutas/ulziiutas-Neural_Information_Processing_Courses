### Homework 3

__2 (b).__ Plot in Python the p.m.f. of X for λ = 1; 3; 5; 50. What do you observe
about the shape of the p.m.f. as λ grows?

--- using definition

```python
def pois_pmf(x,lamda):
    return lamda**x * np.exp(-lamda)/ special.factorial(x)
lambda_all = [1,3,5,50]
x = np.arange(0.,100.,1,)

fig = plt.figure(figsize = (10,6))
for i,lamda in enumerate(lambda_all):
    plt.subplot(2,2,i+1)
    plt.plot(x, pois_pmf(x, lamda),'bo', ms=5,  label='poisson pmf')
    plt.xlabel('x')
    plt.ylabel('probability')
    plt.title(r'$\lambda = $'+str(lamda))

fig.tight_layout()
```

--- using scipy package

```python
lambda_all = [1,3,5,50]
x = np.arange(0.,100.,1)

fig = plt.figure(figsize = (10,6))
for i,lamda in enumerate(lambda_all):
    plt.subplot(2,2,i+1)
    plt.plot(x, stat.poisson.pmf(x, lamda),'bo', ms=5 ,label='poisson pmf')
    plt.xlabel('x')
    plt.ylabel('probability')
    plt.title(r'$\lambda = $'+str(lamda))
    
    
fig.tight_layout()
```

__2 (c).__ What are the parameters of the best matching Normal distribution for
a given Poisson distribution with parameter λ? No formal derivation is required here, use your intuition and verify by plotting (for the same values of
λ as above).

--- Just the intuitive way: mean and variance should be equal
For more detials check this wikipedia article part on "Related distributions": https://en.wikipedia.org/wiki/Poisson_distribution

```python
lambda_all = [1,3,5,50]
x = np.arange(0.,100.,1)

fig = plt.figure(figsize = (12,7))
for i,lamda in enumerate(lambda_all):
    plt.subplot(2,2,i+1)
    plt.plot(x, stat.poisson.pmf(x, lamda),'bo', ms=5 ,label='Poisson')
    plt.plot(x,stat.norm.pdf(x, loc = lamda, scale = np.sqrt(lamda)) ,'r--', ms=5 ,label='Gaussian')
    plt.xlabel('x')
    plt.ylabel('probability')
    plt.title(r'$\lambda = $'+str(lamda))
    plt.legend()

fig.tight_layout()
```
