```python
from IPython.display import display, Math, Latex

# Libraries you might need 
from matplotlib import pyplot as plt
%matplotlib inline
from scipy.stats import bernoulli

import numpy as np
import statistics
import time
from numba import jit, cuda, vectorize
```

```python
N = 10;
f_max = 20;
sigma = 1;
```

## 1-a

```python
def tuning_curve_a(s):
    return f_max * s;
```

```python
def tuning_curve_b(s, j):
    if (j-1)/N < s and s < j/N:
        return N * f_max * (s - (j-1)/N);
    else:
        return 0;
```

```python
s = np.arange(0, 1, 0.001);
f_s_a_5 = tuning_curve_a(s);
f_s_b_5 = np.zeros(len(s));
for i in range(len(s)):
    f_s_b_5[i] = tuning_curve_b(s[i], 5);
```

```python
plt.plot(s, f_s_a_5, label = 'Population A');
plt.plot(s, f_s_b_5, label = 'Population B');
plt.title('Tuning Curve for Neuron 5');
plt.xlabel('s');
plt.ylabel('f(s)');
plt.legend();
```

## 1-b

```python
corr_matrix = np.ones([len(s), len(s)]);

fisher_info_a = np.full(len(s), f_max**2);
fisher_info_b = np.full(len(s), (N*f_max)**2);

fisher_info_b[f_s_b_5 == 0] = 0;
```

```python
plt.figure(figsize=(15, 5));
plt.suptitle('Fisher Information for Neuron 5', fontsize=16);
      
plt.subplot(1, 2, 1);
plt.plot(s, fisher_info_a);
plt.title('Population A');
plt.xlabel('s');
plt.ylabel('I_f(s)');

plt.subplot(1, 2, 2);
plt.plot(s, fisher_info_b);
plt.title('Population B');
plt.xlabel('s');
plt.ylabel('I_f(s)');
```
