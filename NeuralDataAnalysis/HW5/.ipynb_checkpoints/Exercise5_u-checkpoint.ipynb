````python
# this function does not return anything, it just creates a plot! 

def plotRaster(spikes, neuron, scale = 1.0, spacing=1.0, flip=False):
#    Plot spike rasters for a single neuron sorted by condition
#
#    spikes    pandas dataframe with columns 
#                  Neuron
#                  SpikeTimes
#                  Dir
#                  relTime
#                  Trial
#                  stimPeriod
#    neuron    scalar 
    
    if (flip==False):
        flip = 1.0
    else: 
        flip = -1.0
        
    spacing = 1.0 / spacing
            
    
#     plt.figure(figsize=(22, 10))

    # Retrieve the specified neuron's spikes
    neuronSpikes = spikes.where(spikes["Neuron"] == neuron).dropna()  
    
    # Sort the spikes by direction angle & by trial
    neuronSpikes = neuronSpikes.sort_values(['Dir', 'Trial'], ascending=[0,1])   
    
    # Directions
    directions = neuronSpikes["Dir"].unique() #np.sort(stims["Dir"].unique())
    
    # For each direction 
    for k, v in enumerate(directions):
        directedNeuronSpikes = neuronSpikes.where(neuronSpikes["Dir"] == v).dropna() 
        # Use different colors for odd and even 
        color = "red"
        if (k % 2 == 0): 
            color = "blue"
        plt.scatter(directedNeuronSpikes["RelTime"], (directedNeuronSpikes["Trial"]*scale)+flip*spacing*v, c=color, s = 2) 
    
    plt.title(sprintf("Spike Raster plot for Neuron %i", neuron), fontsize=12)
    plt.ylabel("Directions [°]", fontsize=12)
    plt.xlabel("Time [ms]", fontsize=12)
    plt.xticks(fontsize=12)
    plt.yticks(flip*np.arange(0, spacing*np.max(directions), spacing*deltaDir), np.arange(0, np.max(directions), deltaDir), fontsize=12)

#     ax = plt.gca()
#     ax.yaxis.set_major_locator(MultipleLocator(deltaDir))
#     labels = np.arange(0, np.max(directions), deltaDir) 
#     if (flip == -1.0):
#         labels = np.flip(labels)
#     ax.yaxis.set_major_formatter(mpl.ticker.FixedFormatter(labels))
    
#     minorLabels = np.repeat(11, np.shape(directions)[0])
#     ax.yaxis.set_minor_locator(MultipleLocator(np.shape(directions)[0]))
#     ax.yaxis.set_minor_formatter(mpl.ticker.FixedFormatter(minorLabels))
#     ax.tick_params(direction='out', axis="y", which="major", length=10) #, colors='r', grid_color='r', grid_alpha=0.5)

    
````

## Task 3: Fit and plot tuning functions

The goal is to visualize the activity of each neuron as a function of stimulus direction. First, compute the spike counts of each neuron for each direction of motion and trial.  The result should be a matrix `x`, where $x_{jk}$ represents the spike count of the $j$-th response to the $k$-th direction of motion (i.e. each column contains the spike counts for all trials with one direction of motion).	If you used dataframes above, the `groupby()` function allows to implement this very compactely. Make sure you don't loose trials with zero spikes though. Again, other implementations are completely fine.

Fit the tuning curve, i.e. the average spike count per direction, using a von Mises model. To capture the non-linearity and direction selectivity of the neurons, we will fit a modified von Mises function:

$$ f(\theta) = \exp(\alpha + \kappa (\cos (2*(\theta-\phi)))-1) + \nu (\cos (\theta-\phi)-1))$$

Here, $\theta$ is the stimulus direction. Implement the von Mises function in `vonMises()` and plot it to understand how to interpret its parameters $\phi$, $\kappa$, $\nu$, $\alpha$. Perform a non-linear least squares fit using a package/function of your choice. Implement the fitting in `tuningCurve()`. 

Plot the average number of spikes per direction, the spike counts from individual trials as well as your optimal fit.

Select two cells that show nice tuning to test you code.

*Grading: 3 pts*

````python
def computeSpikeCounts(spikes):
    # Construct an artificial dataframe / sql table on which we will perform the join operator and then count the # of spikes -- this deals with 0's
    spikeCountHelper = pd.DataFrame() 
    spikeCountHelper["Neuron"] = np.repeat(np.arange(1, 42, 1), 16*nTrials)
    spikeCountHelper = spikeCountHelper.groupby('Neuron').apply(addDirection)
    spikeCountHelper = spikeCountHelper.groupby(['Neuron', 'Dir']).apply(addTrialNumber) 

    sqlcode = '''
    SELECT spikeCountHelper.Neuron, spikeCountHelper.Dir, spikeCountHelper.Trial, COUNT(spikes.Trial) as SpikeCount
    FROM spikeCountHelper LEFT OUTER JOIN spikes on spikeCountHelper.Neuron = spikes.Neuron and spikeCountHelper.Dir = spikes.Dir and spikeCountHelper.Trial = spikes.Trial
    GROUP BY spikeCountHelper.Neuron, spikeCountHelper.Dir, spikeCountHelper.Trial
    '''

    spikeCounts = psql.sqldf(sqlcode,locals()) 
    return spikeCounts
````

````python
def vonMises(theta, alpha, kappa, nu, phi):
# Evaluate the parametric von Mises tuning curve with parameters p at locations theta. 
# The locations are given in degrees, not radians.
#
#   theta:  1 by N
#   alpha, kappa, nu, phi:   parameters of the function
#
#   f:      1 by N
        
    f = np.exp(alpha + (kappa * np.cos(2*(theta-phi))) - 1) + nu * (np.cos(theta-phi) - 1)
    return f
    
````

````python
def tuningCurve(neuron, dirs, counts, show):
    '''
    Fit a von Mises tuning curve to the spike counts in count with direction dir using a least-squares fit.
    Plot the data if show is True, otherwise just return the fit.
        count      #trials x 1, contains the spike count during the stimulation period
        dirs       #trials x 1, contains the direction in degrees
        show       Boolean, indicates whether a plot is generated

        p          1 by 4 parameter vector of tuning curve function
    '''
    
#     Perform a non-linear least squares fit using a package/function of your choice. Implement the fitting in `tuningCurve()`.  
#     Plot the average number of spikes per direction, the spike counts from individual trials as well as your optimal fit. 
    popt, pcov = opt.curve_fit(vonMises, dirs * np.pi / 180, np.mean(counts, axis=0)) 
        
    if show:
        f = vonMises(dirs * np.pi / 180, popt[0], popt[1], popt[2], popt[3])
        plt.title(sprintf("Optimal Tuning Curve for neuron %i", neuron))
        plt.xlabel("Direction x [degrees]")
        plt.ylabel("Spike Count Per Direction Per Trial")
        plt.plot(dirs, f)
        plt.scatter(np.tile(directions,(11,1)),counts,c='r',s=10)
        return
    else:
        return popt
````

## Task 4: Permutation test for direction tuning

Implement a permutation test to quantitatively assess whether a neuron is direction/orientation selective. To do so, project the vector of average spike counts, $m_k=\frac{1}{N}\sum_j x_{jk}$ on a complex exponential with two cycles, $v_k = \exp(\psi i \theta_k)$, where $\theta_k$ is the $k$-th direction of motion in radians and $\psi \in 1,2$ is the fourier component to test (1: direction, 2: orientation). Denote the projection by $q=m^Tv$. The magnitude $|q|$ tells you how much power there is in the $\psi$-th fourier component. 

Estimate the distribution of |q| under the null hypothesis that the neuron fires randomly across directions by running 1000 iterations where you repeat the same calculation as above but on a random permutation of the trials (that is, randomly shuffle the entries in the spike count matrix x). The fraction of iterations for which you obtain a value more extreme than what you observed in the data is your p-value. Implement this procedure in the function ```testTuning()```. 

Illustrate the test procedure for one of the cells from above. Plot the sampling distribution of |q| and indicate the value observed in the real data in your plot. 

How many cells are tuned at p < 0.01?

Reminder: `x` where $x_{jk}$ represents the spike count of the $j$-th response to the $k$-th direction of motion

````python
def computeMagnitude(neuronSpikeCounts, nu):
    m = np.mean(neuronSpikeCounts, axis=0) 
    q = np.dot(m.T, nu)

    q_magnitude = np.abs(q) 
    return q, q_magnitude


def testTuning(neuron, dirs, neuronSpikeCounts, psi=1, show=False):
    """
    Plot the data if show is True, otherwise just return the fit.
        count      #trials x 1, contains the spike count during the stimulation period
        dirs       #trials x 1, contains the direction in degrees
        psi        fourier component to test (1 = direction, 2 = orientation)
        show       Boolean, indicates whether a plot is generated
    
        p       p-value
        q       magnitude of second Fourier component
        qdistr  sampling distribution of |q| under the null hypothesis    

    """
    
    # Compute nu 
    theta = dirs * np.pi / 180  
    nu = np.exp(psi*1j*theta)
    
    # Retrieve SpikeCounts from Dataframe 
    counts = np.reshape(np.array(neuronSpikeCounts[["SpikeCount"]]), (16, 11)).T  
    
    # Compute |Q| 
    gt_q, gt_q_magnitude = computeMagnitude(counts, nu)
    
    # Create array qdistr which will act as our null hypothesis estimation?
    qdistr = np.zeros(1000)
    q_magnitude_distr = np.zeros(1000)

    for i in range(1, 1001):
        # Shuffle the directions for each Trial, and then retrieve the spike counts from the dataframe
        shuffledCounts = np.reshape(np.array(neuronSpikeCounts.groupby("Trial").apply(shuffle).reset_index(drop=True)[["SpikeCount"]]), (16, 11)).T 
        
        # Compute q and q_magnitude for shuffled data 
        q, q_magnitude = computeMagnitude(shuffledCounts, nu)
        qdistr[i-1] = q
        q_magnitude_distr[i-1] = q_magnitude
    
    p = np.sum(np.array(q_magnitude_distr > gt_q_magnitude, dtype=int)) / 1000 
    
    if show: 
        sns.distplot(q_magnitude_distr)
        plt.title("|q| distribution")
        xMax = np.max(q_magnitude_distr)
        plt.axvline(x=gt_q_magnitude, color="black", linewidth=2, label="GT |q|")           
    else:
        return p, gt_q, gt_q_magnitude, qdistr, q_magnitude_distr

  
````

````python
p_dir = np.zeros(41) 
p_ori = np.zeros(41) 

for i in range(1,42): 
    printf("Iteration: %i\n", i)
    p_dir[i-1], _, _, _, _ = testTuning(i, directions, spikeCounts.where(spikeCounts["Neuron"] == i).dropna(), psi=1, show=False);
    p_ori[i-1], _, _, _, _ = testTuning(i, directions, spikeCounts.where(spikeCounts["Neuron"] == i).dropna(), psi=2, show=False);
````
