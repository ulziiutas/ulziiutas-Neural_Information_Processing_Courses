## 2. Outlier analysis

### E2&3. Show that the same outliers occur in both `Attention` and `Mediation`

```python
# DATA SELECTION

# Initialize that matrix. One for Attention and another for Mediation.
Att = np.zeros((10,10)) # There are 10 subjectes x 10 videos
Med = np.zeros((10,10))
for j in range (10): # j is SubjectID index
    for i in range(10): # i is VideoID index
        Att[j][i] = df['Attention'][j][i]
        Med[j][i] = df['Mediation'][j][i]
        
# Check data:
# Each row corresponds to the values for the same subject in different videos
# Each column corresponds to the values for the same video in different subjects
# print(Att)
# print(Med)
```

```python
# VISUALIZATION
x_array=np.arange(0,10) # Array for the x-axis

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))

# For subjects [j]
for j in range(10):
    # For Attention
    ax1.scatter(x_array, Att[j][:], label='SubjectID {}'.format(j))
    # For Mediation
    ax2.scatter(x_array, Med[j][:], label='SubjectID {}'.format(j))
    
ax1.set_title('Attention of subjects for each video')
ax1.set_xlabel('VideoID')
ax1.set_xticks(np.arange(0, 10, step=1))
ax1.set_ylabel('Attention')
ax1.legend(loc=4)

ax2.set_title('Mediation of subjects for each video')
ax2.set_xlabel('VideoID')
ax2.set_xticks(np.arange(0, 10, step=1))
ax2.set_ylabel('Mediation')
ax2.legend(loc=4)
plt.show()
```

Some outliers can already be visually detected because they differ a lot from the rest of the data. In this case, we consider all data from subject 6 as outlier because it shows at 0, whereas the rest of the data has values around 50. This happens for both Attention and Mediation. We could probably interpret that those zeros as missing values due to some problem in the collection of the data; maybe that subject did not properly wear the EEG cap.

Anyway, we should design an elegant way of defining what an outlier is (better than simple visual inspection). The designed methodology should catch all data from subject 6 and possibly some other data which differs from the trend.

We found two ways of defining outlier:
1. **Boxplot:** It is a method for graphically depicting groups of numerical data through their quartiles. The box extends from the Q1 to Q3 quartile values of the data, with a line at the median (Q2). The whiskers extend from the edges of box to show the range of the data. The position of the whiskers is set by default to 1.5 * IQR (IQR = Q3 - Q1) from the edges of the box. Outlier points are considered those past the end of the whiskers. As it is already built, we did not find any way to customize it so that we could tune it to select more or less outliers.


2. **Z-score:** Determines how many deviations is the data distanced from the mean. This is done by simply applying  $z=\frac{x-u}{\sigma}$. Then, a limit can be selected and any data which exceeds that limit is considered an outlier. It is practical because we can arbitrarly tune this limit to select more or less outliers.

```python
# 1. Boxplot
df.boxplot(column=['Attention', 'Mediation'])
plt.title('Boxplot')
plt.show()
```

By definition, the outliers are represented past the end of the whiskers. For both predictors, we observe the many values (darker color) at zero which correspond again to subject 6. Apart from these, we also detected an outlier for attention with a value much higher than the mean. Checking back the correspondance with the previous representation, we can certainly affirm that it corresponds to that value for subject 2, video 5.

```python
# 2. Z-score
# Compute Z-score
z_att = np.abs(stats.zscore(df['Attention']))
z_med = np.abs(stats.zscore(df['Mediation']))

# Choose limit and find outliers
z_lim = 2 # All values 2 standard deviations further from the mean are considered outliers
Att_out = np.squeeze(np.where(z_att > z_lim))
Med_out = np.squeeze(np.where(z_med > z_lim))
print('Outliers for Attention: {}'.format(Att_out))
print('Outliers for Mediation: {}'.format(Med_out))
```

The correspondance of these outliers to its subject and video can be read in the next way: the first and second number indicate the subject and the video, respectively. So the first outlier for attention, 25, corresponds to subject 2, video 5.

Knowing this, we reaffirm that all data from subject 6 are outliers for both predictors. Adittionaly, we detected another outlier in subject 3, patient 3, which also is present in both predictors. Finally, the value of subject 2, video 5, is also outlier, but only for the predictor Attention.

### E6. Correct skewness

There are different transformations that can be used to correct skewness. Let's check some of them.

```python
# log(x)

skewness_log = np.log(df_clean[predictors]).skew()
print('Skewness after log(x) tranformation:')
print(skewness_log)

# VISUALIZATION
plt.figure(figsize=(10, 8))
plt.suptitle('"Skewness-corrected" with log(x)', fontsize=16)
for i in range(3,11): # Range of columns from Delta to Gamma2
    predictor_i = df_clean.columns[i]
    plt.subplot(4,2,i-3+1)
    np.log(df_clean[predictor_i]).plot.density()
    plt.title(predictor_i)
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.5, wspace=0.35)
```

The skewness improved for most of predictors, except from Delta and Theta. One option could be applying this transformation only to the predictors in which it reduces their skewness.

```python
# sqrt(x)

skewness_sqrt = np.sqrt(df_clean[predictors]).skew()
print('Skewness after sqrt(x) tranformation:')
print(skewness_sqrt)

# VISUALIZATION
plt.figure(figsize=(10, 8))
plt.suptitle('"Skewness-corrected" with sqrt(x)', fontsize=16)
for i in range(3,11): # Range of columns from Delta to Gamma2
    predictor_i = df_clean.columns[i]
    plt.subplot(4,2,i-3+1)
    np.sqrt(df_clean[predictor_i]).plot.density()
    plt.title(predictor_i)
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.5, wspace=0.35)
```

Here, this transformation reduced the skewness for all the predictors excepts from the Delta, which got worse. We could, again, apply it to all predictors with exception of this one.

```python
# Box-Cox
BC = np.zeros((8,90))
plt.figure(figsize=(15, 12))
plt.suptitle('Skewness-corrected distributions', fontsize=16, y=1.02)
for i in range(3,11):
    # Compute Box-Cox
    BC[i-3], _ = stats.boxcox(df_clean[df_clean.columns[i]])
    # skewness-corrected density distributions
    plt.subplot(4,2,i-3+1)
    sns.distplot(BC[i-3])
    plt.title(df_clean.columns[i])
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.7, wspace=0.35)

# How can we compute the skewness for the Box-Cox transformed predictors?
```

All of the distribution seem to have reduced their skewness after the Box-Cox transformation

BoxCox is the transfortmation giving the best corrected skewness, so is the one we use.

```python
# Create new DataFrame df2 with corrected skewness using Boc-Cox transformation 

BC_transposed = np.transpose(BC) # We transpose it so that it has a proper shape
df2 = df_clean.copy()
# Substitute old values for predictors with the ones obtained in BC
predictors_to_correct = df_clean.columns[3:11] # All freq bands
for i in range(len(predictors_to_correct)):
    df2 [predictors_to_correct[i]] = BC_transposed[:,i-3]
    
# df2.head()
```

## 5. Linear regression for prediction

```python
Y = df_clean[predictors]
data_predictors = np.zeros([8, 90])
i = 0
j = 0
for item in Y:
    j = 0
    for columnItems in Y[item]:
        data_predictors[i][j] = columnItems
        j = j + 1
    i = i + 1
data_predictors = data_predictors.transpose()

Y = data_predictors

def loo_function(Y, df_loo):
    loo = LeaveOneOut()
    loo.get_n_splits(Y)
    ytests = []
    ypreds = []

    for train_idx, test_idx in loo.split(Y):
        X_train, X_test = Y[train_idx], Y[test_idx]
        y_train, y_test = df_loo[train_idx], df_loo[test_idx]

        model = sm.OLS(y_train, X_train).fit()
        y_pred = model.predict(X_test)

        ytests += list(y_test)
        ypreds += list(y_pred)

    rr = metrics.r2_score(ytests, ypreds)
    ms_error = metrics.mean_squared_error(ytests, ypreds)
    return ytests, ypreds, rr, ms_error

def k_fol_function(Y, df_loo):
    ytests = []
    ypreds = []

    KFold(n_splits=10, random_state=None, shuffle=False)
    kf = KFold(n_splits=10)
    kf.get_n_splits(Y)


    for train_idx, test_idx in kf.split(Y):    
        X_train, X_test = Y[train_idx], Y[test_idx]
        y_train, y_test = df_loo[train_idx], df_loo[test_idx]

        model = sm.OLS(y_train, X_train).fit()
        y_pred = model.predict(X_test)

        ytests += list(y_test)
        ypreds += list(y_pred)

    rr = metrics.r2_score(ytests, ypreds)
    ms_error = metrics.mean_squared_error(ytests, ypreds)
    return ytests, ypreds, rr, ms_error
```

```python
df_clean['Mediation'] = list(map(int, df_clean['Mediation']))
df_clean['Attention'] = list(map(int, df_clean['Attention']))

mediation_tests, mediation_preds, mediation_rr, mediation_ms_error = loo_function(Y, df_clean['Mediation'])
attention_tests, attention_preds, attention_rr, attention_ms_error = loo_function(Y, df_clean['Attention'])

print("Leave One Out Cross Validation")
print("Mediation R^2: {:.5f}%, MSE: {:.5f}".format(mediation_rr*100, mediation_ms_error))
print("Attention R^2: {:.5f}%, MSE: {:.5f}".format(attention_rr*100, attention_ms_error))

plt.figure(figsize=(20, 8))

plt.suptitle('Mediation and Attention Predictors Leave One Out', fontsize=16)
plt.subplot(1,2,1)
plt.scatter(mediation_tests, mediation_preds)
plt.title("Mediation Prediction for Untransformed Data(Leave One Out CV)")
plt.xlabel(r'Mediation')
plt.ylabel(r'Prediction')

plt.subplot(1,2,2)
plt.scatter(attention_tests, attention_preds)
plt.title("Attention Prediction for Untransformed Data(Leave One Out CV)")
plt.xlabel(r'Attention')
plt.ylabel(r'Prediction')
plt.subplots_adjust(top=0.89, bottom=0.08, left=0.05, right=0.95, hspace=0.40, wspace=0.35)
plt.show()
```

```python

mediation_kfold_tests, mediation_kfold_preds, mediation_kfold_rr, mediation_kfold_mse = k_fol_function(Y, df_clean['Mediation'])

attention_kfold_tests, attention_kfold_preds, attention_kfold_rr, attention_kfold_mse = k_fol_function(Y, df_clean['Attention'])

print("K-Fold Cross Validation")
print("Mediation R^2: {:.5f}%, MSE: {:.5f}".format(mediation_kfold_rr, mediation_kfold_mse))
print("Attention R^2: {:.5f}%, MSE: {:.5f}".format(attention_kfold_rr*100, attention_kfold_mse))

plt.figure(figsize=(20, 8))

plt.suptitle('Mediation and Attention Predictors K-Fold', fontsize=16)
plt.subplot(1,2,1)
plt.scatter(mediation_kfold_tests, mediation_kfold_preds)
plt.title("Mediation Prediction for Untransformed Data(K-Fold CV)")
plt.xlabel(r'Mediation')
plt.ylabel(r'Prediction')

plt.subplot(1,2,2)
plt.scatter(attention_kfold_tests, attention_kfold_preds)
plt.title("Attention Prediction for Untransformed Data(K-Fold CV)")
plt.xlabel(r'Attention')
plt.ylabel(r'Prediction')
plt.subplots_adjust(top=0.89, bottom=0.08, left=0.05, right=0.95, hspace=0.40, wspace=0.35)
plt.show()
```

We have used two Cross Validation which are K-Fold and Leave One Out(LOO). As can be seen above graphs, the untransformed data is too noisy and causes really bad model so that the R-squared and MSE is showing terrible scores. However, we can see that Mediation shows better results than Attention as we have shown from the previous exercise.
