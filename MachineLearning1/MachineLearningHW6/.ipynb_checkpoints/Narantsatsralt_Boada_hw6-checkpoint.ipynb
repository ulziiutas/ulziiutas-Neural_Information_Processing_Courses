```python
# Function to plot tables
class color:
   BOLD = '\033[1m'
   END = '\033[0m'
  
def print_table(table):
    longest_cols = [
        (max([len(str(row[i])) for row in table]) + 3)
        for i in range(len(table[0]))
    ]
    row_format = "".join(["{:>" + str(longest_col) + "}" for longest_col in longest_cols])
    i = 0
    for row in table:
        if i == 0:
            print(color.BOLD + row_format.format(*row) + color.END)
            print('------------------------------------------------------------------------------------------------------------------')
        else:
            print(row_format.format(*row))
        i = i +1
```

By summing the frequency powers of all the bands in the respective block:

* **`LowFreqPower`** = `Delta` + `Theta` + `Alpha1` + `Alpha2` + `Beta1`

```python
# Define predictors groups
predictors_LowFreq = df.columns[3:8]
predictors_HighFreq = df.columns[8:11]

# Create new predictors LowFreqPower and HighFreqPower
df = df.assign(LowFreqPower = lambda df: df[predictors_LowFreq].sum(axis=1))
df = df.assign(HighFreqPower = lambda df: df[predictors_HighFreq].sum(axis=1))

# Check them
df.head()
```

#### ---------- Our K-means ----------

```python
# HELPER: Euclidean Distance Calculator
def dist(a, b, ax=1):
    # a, b: Points between which we compute the distance. They must by np.matrix()
    return np.linalg.norm(a - b, axis=ax)
```

##### Decide K and initialize centroids

```python
# Define # of clusters
K = 2

# Initialize centroids (randomly)
np.random.seed(0)
# X coordinates of random centroids
C_x = np.random.randint(np.min(X[:,0]), np.max(X[:,0]), size=K) # generates random number within interval
# Y coordinates of random centroids
C_y = np.random.randint(np.min(X[:,1]), np.max(X[:,1]), size=K)

# Merge centroids coordinates in one variable, C
C_homemade = np.array(list(zip(C_x, C_y)), dtype=np.float32)
print('Our initial centroids:\n {}'.format(C_homemade))

# Plot centroids
plt.scatter(LowFreq, HighFreq, c='k', s=7, alpha=0.6)
plt.scatter(C_x[0], C_y[0], marker='*', s=200, c='b', label='C-1') # Centroid 1
plt.scatter(C_x[1], C_y[1], marker='*', s=200, c='r', label='C-2') # Centroid 2
plt.axis('equal')
plt.xlabel('f1: LowFreqPower')
plt.ylabel('f1: HighFreqPower')
plt.legend()
plt.title('Observations with randomly initialized centroids')

plt.plot()
```

##### Expectation Maximization (EM) 

```python
# C_old: saves the old centroids, to monitor the improvement at each iteration
C_old = np.zeros(C_homemade.shape)
# labels: contains the label for each observation
labels_homemade = np.zeros(len(X))
# Error: distance between new centroids and old centroids
error = dist(C_homemade, C_old, None)

# While-loop runs until the error = 0, i.e. centroids have NOT moved, i.e. algorithm has converged
while error != 0:
    
    # E-Step: Re-assing labels of every observation
    for i in range(len(X)):
        distances = dist(X[i], C_homemade) # returns the distances from the point X[i] to each centroids
        labels_homemade[i] = np.argmin(distances) # Label the point with the label of the closest centroid
        
    # Storing the old centroid values
    C_old = C_homemade
    
    # M-Step: Re-compute centroids
    for i in range(K):
        # From all observations, take all points belonging to Cluster i
        points = [X[j] for j in range(len(X)) if labels_homemade[j] == i]
        # Use those points to calculate their centroid
        C_homemade[i] = np.mean(points, axis=0)
        
    # Calculate how much did the centroids move
    error = dist(C_homemade, C_old, None)
```

#### ---------- K-means using sklearn ----------
Much easier because everything is already implemented.

```python
# Initialize model
kmeans = KMeans(n_clusters=K, random_state=0)
# Train it with our data
kmeans = kmeans.fit(X)

# OUTPUT
# Centroid coordinates
C_sklearn = kmeans.cluster_centers_
# Cluster labels
labels_sklearn = kmeans.predict(X)
```

### E3. Transformation of the 2D predictors

K-means calculates the distances using the Euclidean norm, this is why it may not take much information of one predictor (i.e. HighFreqPower) if the other predictor with which we are comparing is spread in a much bigger range (i.e. LowFreqPower). Scaling the can solve this problem, because then all predictors will be in the same numerical scale range. 

#### Scaling
A typical way of scaling is subtracting the mean and dividing by the standard deviation. We will add 2 new columns in our dataframe df with the scaled predictors. We scale for all the subjects at the same time. i.e. mean = sum(all the column)/90.

```python
# Define new predictors which we want to scale
predictors_new = ['LowFreqPower', 'HighFreqPower']
predictors_scaled = ['LFscaled', 'HFscaled']

for i in range(len(predictors_scaled)):
    baseline_i = df[predictors_new[i]].mean()
    SD_i = df[predictors_new[i]].std()
    df[predictors_scaled[i]] = df[predictors_new[i]].subtract(baseline_i).divide(SD_i)

df.head()
```

```python
# Get their values out from df
LFscaled = df['LFscaled'].values
HFscaled = df['HFscaled'].values

# Merge them in X, which then stores the coordinates of each observation
X_scaled = np.array(list(zip(LFscaled, HFscaled)))

# Plot

fig, ax = plt.subplots(1,2,figsize=(15,4))

ax[1].scatter(LFscaled, HFscaled, c='k', s=7)
ax[0].scatter(LowFreq, HighFreq, c='k', s=7)

# Plot x and y axis at the same scale
ax[1].axis('equal')
ax[0].axis('equal')

ax[1].set_xlabel('LFscaled')
ax[1].set_ylabel('HFscaled')
ax[1].set_title('Observations in LFscaled vs HFscaled')

ax[0].set_xlabel('LowFreqPower')
ax[0].set_ylabel('HighFreqPower')
ax[0].set_title('Observations in LowFreqPower vs HighFreqPower')

plt.plot()
```

When plotting both using the same scaling of x and y axes, we see how the distances have been scaled, hence, K-means is expected to do a better job. Previously, K-means was practically only being affected by distanes in the x-axis, now, it will also take profit of the distance in the y-axis.

```python
# Initialize model
kmeans = KMeans(n_clusters=K, random_state=0)
# Train it with our data
kmeans = kmeans.fit(X_scaled)

# OUTPUT
# Centroid coordinates
C_scaled = kmeans.cluster_centers_
# Cluster labels
labels_scaled = kmeans.predict(X_scaled)
```

In order to properly understand how ROC curve and AUC are calculated, we firstly execute logistic regression using all data for training and that same data to evaluate. Later, we will be ready to face how to plot ROC and calculate AUC for logistic regression evauluated in repeated K-folds crossvalidation, which will give a more robust evaluation.

#### Simple logistic regression
To get an idea of how to execute logistic regression using sklearn, plot ROC and calculate AUC. 

```python
# Set predictors and target
predictors = df.columns[3:11]
target = 'ReportedConfusion'

# Initialize model
logreg = LogisticRegression(penalty='l2', C=100) # C is inverse of regulatization strength

# Train model with our data
logreg.fit(df[predictors], df[target])
```

Once the model is trained, we can start classifying new observations or evaluating our model by classifying observation from which we know their real class, i.e. the ones we used to train the model.

Remember that logistic regression returns the probability of an observation to belong to a certain class. Then, we need to set a threshold at which we classify an observation from either one or the other class. To get the probabilites and classify for a threshold at 0.5, we use the next functions:

```python
# For all observations used in the training

# Probabilities
probs = logreg.predict_proba(df[predictors]) # returns a Nx2 array. Column 1 = P(class1), Column 2 = P(Class2). Hint: They add to 1!
# Classification w/ threshold = 0.5
labels_logreg = logreg.predict(df[predictors]);

# Check that the threshold is really at 0.5
# print(probs)
# print(len(labels_logreg))
```

Once we classified by setting a specific threshold, we can calculate the confusion matrix and some evaluation parameters.

```python
# Confusion matrix
confusion_matrix(labels_logreg, labels_real)
# Extract tp, fp, tn, fn from the confusion matrix
tn, fp, fn, tp = cnf_matrix.ravel()

table = [
    ["", "Confused ", "Not confused "],
    ["Predicted as confused", tp, fp],
    ["Predicted as not confused", fn, tn]
]
print_table(table)
# Accuracy
acc = (tn+tp)/(tn+tp+fn+fp)
# Sensitivity
sens = tp / (tp+fn)
# Specificity
spec = tn / (tn+fp)

print('\nAccuracy: {}\nSensitivity: {} \nSpecificity: {}'.format(round(acc,3), round(sens,3), round(spec,3)))
```

To plot the ROC, we need to get the value of all possible threshold giving rise to different classifications, and from those, calculate the sensitivity and specificity. The function roc_curve(y_true, y_score, pos_label) does the work for us.

* y_true: Real labels
* y_score: Probabilites given by the model
* pos_label: Class at which those probabilities correspond.

And it returns:

* fpr: False positive rate for each different threshold.
* tpr: True positive rate for each different threshold.
* threshold: All different threshold giving rise to different classifications.

Using those outputs, we can plot the ROC

```python
# Calculate parameters for ROC
fpr, tpr, thresholds = roc_curve(labels_real, probs[:,1], pos_label=1) # probs[:,1] are the probabilities of 'confused' for all observations

# Calculate AUC
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(5, 5))
lw = 2
# ROC curve
plt.plot(fpr, tpr, color='darkorange',
         lw=lw, label='ROC curve (area = %0.2f)' % roc_auc)
# Reference diagonal
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC using training set')
plt.legend(loc="lower right")
plt.show()
```

Note that as we are evauluatin using the same dat that we use to train, we get a very good classification for most of the different thresholds. This is expressed in the AUC, which is very near to 1. If we were testing the model in new data that wasn't used for training, we would get a rather lower AUC.

This is how an inividual ROC from one iteration of our repeated K-fold CV will look like.

#### Logistic regression with repeated K-folds cross validation
Repeated K-folds cross validation is a good tool to get a robust evaluation of our model.

```python
# initialize variables
probs_array = [] # here we'll save probabilities of the test data
y_testset_array = [] # here we'll save the real labels in the correct order so that they can be correlated with y_scores. They actually come from the ReportedConfusion
predset_array = [] # prediction set
acc_array = [] # Here we'll save each accuracies obtained for each cross-validation

base_fpr = np.linspace(0, 1, 100) # for the mean ROC
tprs_array = [] # for the mean ROC
AUC_array = []
# split the data in all possible folds, 10 times (each with differently shuffled data)
n_splits = 9
n_repeats = 10
kf = RepeatedKFold(n_splits, n_repeats, random_state=0)

for train_indices, test_indices in kf.split(df[predictors],df[target]):
    
     # gather training set data
    x_trainset = df[predictors].iloc[train_indices]
    y_trainset = df[target].iloc[train_indices]
    
    # gather test set data
    x_testset = df[predictors].iloc[test_indices]
    y_testset = df[target].iloc[test_indices]

    # initialize model
    model = LogisticRegression() # Shouldnt we set penalty to 'none' or equivalently 'l2' and C = very big ???
    
    # train the model
    train_fit = model.fit(x_trainset, y_trainset)
    
    # FOR ACCURACY
    
    # save real labels from test set correctly ordered
    y_testset_array.append(y_testset)
    
    # prediction array
    predset_array.append(model.predict(x_testset))
    
    # get accuracy of each CV iteration and save it in acc_array
    acc_array.append(model.score(x_testset, y_testset))
    
    # FOR INDIVIDUAL AND AVERAGE ROC & AUC
    
    # predict on the test set (get probabilities)
    probs = model.predict_proba(x_testset)
    # calculate parameters for ROC
    
    fpr, tpr, _ = roc_curve(y_testset, probs[:,1], pos_label=1) # probs[:,1] are the probabilities of 'confused' for all observations
    # plot ROC
    plt.plot(fpr, tpr, color='orange', lw=lw, alpha=0.3) # ,label='ROC curve (AUC = {})'.format(round(AUC,2)))

    # calculate AUC and save it in AUC_array
    AUC = auc(fpr, tpr)
    AUC_array.append(AUC)
    
    # For the average ROC
    tprs = interp(base_fpr, fpr, tpr) 
    tprs[0] = 0.0 # set 1st value from iterpolation to zero
    tprs_array.append(tprs)

# AVERAGE ROC
tprs_array = np.array(tprs_array)
mean_tprs = tprs_array.mean(axis=0)

# For std area
std = tprs_array.std(axis=0)
tprs_upper = np.minimum(mean_tprs + std, 1)
tprs_lower = mean_tprs - std

plt.plot(base_fpr, mean_tprs, 'r', label='Average ROC') # average ROC
plt.fill_between(base_fpr, tprs_lower, tprs_upper, color='grey', alpha=0.3) # std area
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--') # reference diagonal
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Logistic Regression (RepeatedKFold)')
plt.legend(loc="lower right")
plt.show()

# ACCURACY
acc_logreg_mean = np.mean(acc_array)
acc_logreg_SD = np.std(acc_array)
print('Accuracy:\n   mean = {}\n   SD = {}\n'.format(round(acc_logreg_mean,3), round(acc_logreg_SD,3)))

# AUC
AUC_logreg_mean = np.mean(AUC_array)
AUC_logreg_SD = np.std(AUC_array)
print('AUC:\n   mean = {}\n   SD = {}'.format(round(AUC_logreg_mean,3), round(AUC_logreg_SD,3)))
```

Each orange line corresponds to the ROC from a single iteration of the CV. As we do not have much data, the test sets contain only 10 data points. This is why the "resolution" of the orange lines is very low.

The red line is the average of all orange ROCs. We interpolated the ROC of each iteration to get the same amount of points and then averaged (it's quite complex to explain it, sorry).

```python
def auroc_func(y, score, k_repeat):
        
    fpr_arr = []
    tpr_arr = []
    auroc = []
    
    for i in range(100):
        fpr, tpr, _ = roc_curve(y[i * k_repeat:(i+1)*k_repeat], score[i * k_repeat:(i+1)*k_repeat], pos_label=1)
        
        if str(auc(fpr, tpr)) == 'nan':
            auroc.append(0)            
        else:            
            auroc.append(auc(fpr, tpr))
        
    return auroc
```

```python
kf = RepeatedKFold(n_splits=10, n_repeats=10)

l1_c_array = np.logspace(2, 3, 100)
l2_c_array = np.logspace(1, 10, 100)

test_tar = []

l2_test_pre = []
l1_test_pre = []

l1_test_acc = []
l2_test_acc = []

l1_test_mrse = []
l2_test_mrse = []

l1_coef = []
l2_coef = []

for i in range(len(l1_c_array)):    
    l1_test_acc.append([])    
    l2_test_acc.append([])    
    
    l1_test_mrse.append([])    
    l2_test_mrse.append([])  
    
    l1_coef.append([])
    l2_coef.append([])
    
    l1_test_pre.append([])
    l2_test_pre.append([])

for train_index, test_index in kf.split(X=df[predictors],y=df[target]):
    
    training_predictors = df[predictors].iloc[train_index]
    training_targets = df[target].iloc[train_index]

    test_predictors = df[predictors].iloc[test_index]
    test_targets = df[target].iloc[test_index]
    
    t =[]    
    
    for i in test_targets:
        t.append(i)
            
    test_tar.append(t)
        
    for i in range(len(l1_c_array)):
        
        clf_l1_LR = Lasso()
        clf_l2_LR = Ridge()
        
        clf_l1_LR.set_params(alpha = l1_c_array[i])
        clf_l2_LR.set_params(alpha = l2_c_array[i])
        
        clf_l1_LR.fit(training_predictors,training_targets)
        clf_l2_LR.fit(training_predictors,training_targets)
        
        scor1 = clf_l1_LR.score(test_predictors,test_targets)
        scor2 = clf_l2_LR.score(test_predictors,test_targets)
                
        pred1 = clf_l1_LR.predict(test_predictors)
        pred2 = clf_l2_LR.predict(test_predictors)
                
        l1_test_pre[i].append(pred1)
        l2_test_pre[i].append(pred2)
                
        mrse1 = (pred1 - t) ** 2
        mrse2 = (pred2 - t) ** 2
        
        l1_test_mrse[i].append(mrse1)
        l2_test_mrse[i].append(mrse2)
        
        l1_coef[i].append(clf_l1_LR.coef_)
        l2_coef[i].append(clf_l2_LR.coef_)
        
        l1_test_acc[i].append(scor1)
        l2_test_acc[i].append(scor2)
                        
print('Finished')
```

```python
mrse_l2 = []
coef_l2 = []

acc_mean_l2 = []
acc_sd_l2 = []

l2_log = np.linspace(1, 10, 100)

for i in range(len(l2_c_array)):
    mrse_l2.append(np.sqrt(np.sum(l2_test_mrse[i])/(len(l2_test_mrse[i]) * len(l2_test_mrse[i][0]))))
    coef_l2.append(np.mean(l2_coef[i]))
    
    acc_mean_l2.append(np.mean(l2_test_acc[i]))
    acc_sd_l2.append(statistics.stdev(l2_test_acc[i]))

optimal_c_value_ridge = round(l2_log[mrse_l2.index(np.min(mrse_l2))], 3)
acc_logreg_ridge_mean = round(acc_mean_l2[mrse_l2.index(np.min(mrse_l2))], 3)
acc_logreg_ridge_sd = round(acc_sd_l2[mrse_l2.index(np.min(mrse_l2))], 3)

auroc_ridge = auroc_func(np.array(test_tar).ravel(), np.array(l2_test_pre[mrse_l2.index(np.min(mrse_l2))]).ravel(), 9)
auroc_logreg_ridge_mean = round(np.mean(auroc_ridge), 3)
auroc_logreg_ridge_sd = round(statistics.stdev(auroc_ridge), 3)

plt.figure(figsize=(20, 16))
plt.suptitle('L2(Ridge) Penalty', fontsize=30)

plt.subplot(2,2,1)    
plt.plot(l2_c_array, mrse_l2)
plt.title('L2-penalized(Ridge) RMSE-Penalty plot', fontsize=22)
plt.xlabel(r'Penalty', fontsize=16)
plt.ylabel(r'RMSE', fontsize=16)

plt.subplot(2,2,2)
plt.plot(l2_c_array, coef_l2)
plt.title('L2-penalized(Ridge) Standard Coefficient-Penalty plot', fontsize=22)
plt.xlabel(r'Penalty', fontsize=16)
plt.ylabel(r'Standard Coefficient', fontsize=16)
plt.show()

plt.subplots_adjust(top=0.89, bottom=0.08, left=0.05, right=0.95, hspace=0.40, wspace=0.35)
plt.show()

print('Optimal Value for L2 Penalty is:', optimal_c_value_ridge)
print('\nRMSE For L2 Penalty(penalty of {}): {}'.format(optimal_c_value_ridge, np.min(mrse_l2)))
print('\nAccuracy Mean For L2 Penalty(penalty of {}): {}'.format(optimal_c_value_ridge, acc_logreg_ridge_mean))
print('\nAccuracy SD For L2 Penalty(penalty of {}): {}'.format(optimal_c_value_ridge, acc_logreg_ridge_sd))

print('\nAUROC Mean For L2 Penalty(penalty of {}): {}'.format(optimal_c_value_ridge, auroc_logreg_ridge_mean))
print('\nAUROC SD For L2 Penalty(penalty of {}): {}'.format(optimal_c_value_ridge, auroc_logreg_ridge_sd))
```
