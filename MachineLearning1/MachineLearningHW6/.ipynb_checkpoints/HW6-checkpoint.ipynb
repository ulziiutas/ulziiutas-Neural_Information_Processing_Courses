```python
class color:
   BOLD = '\033[1m'
   END = '\033[0m'
  
def print_table(table):
    longest_cols = [
        (max([len(str(row[i])) for row in table]) + 3)
        for i in range(len(table[0]))
    ]
    row_format = "".join(["{:>" + str(longest_col) + "}" for longest_col in longest_cols])
    i = 0
    for row in table:
        if i == 0:
            print(color.BOLD + row_format.format(*row) + color.END)
            print('------------------------------------------------------------------------------------------------------------------')
        else:
            print(row_format.format(*row))
        i = i +1
```

```python
%matplotlib inline
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

datapath = 'EEG_data.csv'
rawdata = pd.read_csv(datapath)

rawdata['SubjectID'] = rawdata['SubjectID'].astype('int')
rawdata['VideoID'] = rawdata['VideoID'].astype('int')

rawdata.columns = rawdata.columns.str.replace('predefinedlabel','ExpectedConfusion')
rawdata.columns = rawdata.columns.str.replace('user-definedlabeln','ReportedConfusion')

rawdata['ReportedConfusion'] = rawdata['ReportedConfusion'].astype('int')
rawdata['ExpectedConfusion'] = rawdata['ExpectedConfusion'].astype('int')

data = rawdata.copy()
data = data.groupby(['SubjectID','VideoID']).median()
data = data.drop([6])
data_bc_raw = data.copy()
data_bc = pd.DataFrame()

predictors = ['Delta', 'Theta', 'Alpha1', 'Alpha2', 'Beta1', 'Beta2', 'Gamma1', 'Gamma2']

for (SubjectID), subjectdata in data_bc_raw.groupby(('SubjectID')):
    for i in predictors:
        mean = np.mean(subjectdata[i]) # then by predictor
        subjectdata[i] = subjectdata[i] - mean
    data_bc = pd.concat([data_bc,subjectdata],ignore_index=True)

print(data_bc.shape)
```

```python
# Function for REPEATED CROSS-VALIDATION

def model_repeatedKfolds(predictors, target, K = 10, n_repeats = 10):

    kf = RepeatedKFold(n_splits=K, n_repeats=n_repeats)
    
    test_acc = []
    test_tar = []
    test_pre = []
    test_dec = []
    
    for train_index, test_index in kf.split(X=predictors,y=target):

        model = linear_model.LogisticRegression()
            
        training_predictors = predictors.iloc[train_index]
        training_targets = target.iloc[train_index]


        test_predictors = predictors.iloc[test_index]
        test_targets = target.iloc[test_index]
        
        decision = model.fit(training_predictors,training_targets).decision_function(test_predictors)
        
        pred = model.predict(test_predictors)
        
        test_tar.append(test_targets)        
        
        test_pre.append(pred)        
        
        test_dec.append(decision)
        
        test_acc.append(model.score(test_predictors,test_targets))

    mean_test = np.mean(test_acc)
    sd_test = statistics.stdev(test_acc)


    print('Accuracy of Logistic Regression cross validation with {} folds on test data:'.format(K))
    print('\tmean = {}'.format(mean_test))
    print('\tstandard deviation = {}'.format(sd_test))
    
    return mean_test, sd_test, test_tar, test_pre, test_dec, test_acc
```

```python
print('Logistic regression on baseline corrected data')
log_test_mean, log_test_sd, test_tar, test_pre, test_dec, test_acc = model_repeatedKfolds(predictors=data_bc[predictors], target=data_bc['ReportedConfusion'])
```

```python
low_predictors = ['Delta', 'Theta', 'Alpha1', 'Alpha2', 'Beta1']
high_predictors = ['Beta2', 'Gamma1', 'Gamma2']

LowFreqPower = np.zeros(90)
HighFreqPower = np.zeros(90)

for l_predictor in low_predictors:
    for i in range(len(data_bc[l_predictor])):        
        LowFreqPower[i] = LowFreqPower[i] + data_bc[l_predictor][i]
        
for h_predictor in high_predictors:
    for i in range(len(data_bc[high_predictors])):        
        HighFreqPower[i] = HighFreqPower[i] + data_bc[h_predictor][i]
```

```python
def dist(x1, x2, y1, y2):
    return np.sqrt(np.power((x1 - x2), 2) + np.power((y1 - y2), 2))
```

```python
def k_means(LowFreqPower, HighFreqPower):
    center_1 = random.randint(0,89)
    center_2 = random.randint(0,89)

    center_1_pos_low = LowFreqPower[center_1]
    center_1_pos_high = HighFreqPower[center_1]

    center_2_pos_low = LowFreqPower[center_2]
    center_2_pos_high = HighFreqPower[center_2]

    classification = np.zeros(90)

    for i in range(90):
        dist_1 = dist(LowFreqPower[i], center_1_pos_low, HighFreqPower[i], center_1_pos_high)
        dist_2 = dist(LowFreqPower[i], center_2_pos_low, HighFreqPower[i], center_2_pos_high)

        if dist_1 < dist_2:
            classification[i] = 1
        else:
            classification[i] = 0
    cost_new = 100
    j = 0
    while 0 != cost_new and j < 1000:

        center_1_sum_low = 0
        center_1_sum_high = 0

        c1_len = 0

        center_2_sum_low = 0
        center_2_sum_high = 0

        c2_len = 0

        for i in range(90):

            center_1_sum_low = center_1_sum_low + LowFreqPower[i] * classification[i]
            center_1_sum_high = center_1_sum_high + HighFreqPower[i] * classification[i]


            center_2_sum_low = center_2_sum_low + LowFreqPower[i] * (1 - classification[i])
            center_2_sum_high = center_2_sum_high + HighFreqPower[i] * (1 - classification[i])

            if classification[i] == 1:
                c1_len = c1_len + 1
            else:            
                c2_len = c2_len + 1

        center_1_pos_low = center_1_sum_low / c1_len
        center_1_pos_high = center_1_sum_high / c1_len

        center_2_pos_low = center_2_sum_low / c2_len
        center_2_pos_high = center_2_sum_high / c2_len

        error_val = 0

        for i in range(90):
            dist_1 = dist(LowFreqPower[i], center_1_pos_low, HighFreqPower[i], center_1_pos_high)
            dist_2 = dist(LowFreqPower[i], center_2_pos_low, HighFreqPower[i], center_2_pos_high)

            if dist_1 < dist_2:
                if classification[i] != 1:
                    error_val = error_val + 1
                classification[i] = 1
            else:
                if classification[i] != 0:
                    error_val = error_val + 1
                classification[i] = 0
        cost_new = error_val
        j = j + 1
        print('Error from the last iteration: ', error_val)
        
    return classification, center_1_pos_low, center_1_pos_high, center_2_pos_low, center_2_pos_high
```

```python
classification, center_1_pos_low, center_1_pos_high, center_2_pos_low, center_2_pos_high = k_means(LowFreqPower, HighFreqPower)
```

```python
def roc_auc(y, score, k_repeat):
    
    fpr_arr = []
    tpr_arr = []
    roc_auc = []
    
    tprs = []
    base_fpr = np.linspace(0, 1, 100)
    for i in range(100):
        fpr, tpr, _ = roc_curve(y[i * k_repeat:(i+1)*k_repeat], score[i * k_repeat:(i+1)*k_repeat], pos_label=1)
        
        roc_auc.append(auc(fpr, tpr))
        fpr_arr.append(fpr)
        tpr_arr.append(tpr)        
        
        tpr = np.interp(base_fpr, fpr, tpr)
        tpr[0] = 0.0
        tprs.append(tpr)
        
        
    return roc_auc, fpr_arr, tpr_arr, tprs
```

## E6

```python
def auroc_func(y, score, k_repeat):
        
    fpr_arr = []
    tpr_arr = []
    auroc = []
    
    for i in range(100):
        fpr, tpr, _ = roc_curve(y[i * k_repeat:(i+1)*k_repeat], score[i * k_repeat:(i+1)*k_repeat], pos_label=1)
        
        if str(auc(fpr, tpr)) == 'nan':
            auroc.append(0)            
        else:            
            auroc.append(auc(fpr, tpr))
        
    return auroc
```

```python
kf = RepeatedKFold(n_splits=10, n_repeats=10)

l1_c_array = np.logspace(2, 3, 100)
l2_c_array = np.logspace(1, 10, 100)

test_tar = []

l2_test_pre = []
l1_test_pre = []

l1_test_acc = []
l2_test_acc = []

l1_test_mrse = []
l2_test_mrse = []

l1_coef = []
l2_coef = []

for i in range(len(l1_c_array)):    
    l1_test_acc.append([])    
    l2_test_acc.append([])    
    
    l1_test_mrse.append([])    
    l2_test_mrse.append([])  
    
    l1_coef.append([])
    l2_coef.append([])
    
    l1_test_pre.append([])
    l2_test_pre.append([])

for train_index, test_index in kf.split(X=data_bc[predictors],y=data_bc['ReportedConfusion']):
    
    training_predictors = data_bc[predictors].iloc[train_index]
    training_targets = data_bc['ReportedConfusion'].iloc[train_index]

    test_predictors = data_bc[predictors].iloc[test_index]
    test_targets = data_bc['ReportedConfusion'].iloc[test_index]
    
    test_tar.append(test_targets)
        
    for i in range(len(l1_c_array)):
        
        clf_l1_LR = Lasso()
        clf_l2_LR = Ridge()
        
        clf_l1_LR.set_params(alpha = l1_c_array[i])
        clf_l2_LR.set_params(alpha = l2_c_array[i])
        
        clf_l1_LR.fit(training_predictors,training_targets)
        clf_l2_LR.fit(training_predictors,training_targets)
        
        scor1 = clf_l1_LR.score(test_predictors,test_targets)
        scor2 = clf_l2_LR.score(test_predictors,test_targets)
                
        pred1 = clf_l1_LR.predict(test_predictors)
        pred2 = clf_l2_LR.predict(test_predictors)
                
        l1_test_pre[i].append(pred1)
        l2_test_pre[i].append(pred2)
                
        mrse1 = (pred1 - test_targets) ** 2
        mrse2 = (pred2 - test_targets) ** 2
        
        l1_test_mrse[i].append(mrse1)
        l2_test_mrse[i].append(mrse2)
        
        l1_coef[i].append(clf_l1_LR.coef_)
        l2_coef[i].append(clf_l2_LR.coef_)
        
        l1_test_acc[i].append(scor1)
        l2_test_acc[i].append(scor2)
                        
print('Finished')
```

```python
mrse_l2 = []
coef_l2 = []

acc_mean_l2 = []
acc_sd_l2 = []

l2_log = np.linspace(1, 10, 100)

for i in range(len(l2_c_array)):
    mrse_l2.append(np.sqrt(np.sum(l2_test_mrse[i])/(len(l2_test_mrse[i]) * len(l2_test_mrse[i][0]))))
    coef_l2.append(np.mean(l2_coef[i]))
    
    acc_mean_l2.append(np.mean(l2_test_acc[i]))
    acc_sd_l2.append(statistics.stdev(l2_test_acc[i]))

optimal_c_value_ridge = round(l2_log[mrse_l2.index(np.min(mrse_l2))], 3)
E6_acc_logreg_ridge_mean = round(acc_mean_l2[mrse_l2.index(np.min(mrse_l2))], 3)
E6_acc_logreg_ridge_sd = round(acc_sd_l2[mrse_l2.index(np.min(mrse_l2))], 3)

auroc_ridge = auroc_func(np.array(test_tar).ravel(), np.array(l2_test_pre[mrse_l2.index(np.min(mrse_l2))]).ravel(), 9)
E6_auroc_logreg_ridge_mean = round(np.mean(auroc_ridge), 3)
E6_auroc_logreg_ridge_sd = round(statistics.stdev(auroc_ridge), 3)

plt.figure(figsize=(20, 16))
plt.suptitle('L2(Ridge) Penalty', fontsize=30)

plt.subplot(2,2,1)    
plt.plot(l2_c_array, mrse_l2)
plt.title('L2-penalized(Ridge) RMSE-Penalty plot', fontsize=22)
plt.xlabel(r'Penalty', fontsize=16)
plt.ylabel(r'RMSE', fontsize=16)

plt.subplot(2,2,2)
plt.plot(l2_c_array, coef_l2)
plt.title('L2-penalized(Ridge) Standard Coefficient-Penalty plot', fontsize=22)
plt.xlabel(r'Penalty', fontsize=16)
plt.ylabel(r'Standard Coefficient', fontsize=16)
plt.show()

plt.subplots_adjust(top=0.89, bottom=0.08, left=0.05, right=0.95, hspace=0.40, wspace=0.35)
plt.show()

print('Optimal Value for L2 Penalty is:', optimal_c_value_ridge)
print('\nRMSE For L2 Penalty(penalty of {}): {}'.format(optimal_c_value_ridge, np.min(mrse_l2)))
print('\nAccuracy Mean For L2 Penalty(penalty of {}): {}'.format(optimal_c_value_ridge, E6_acc_logreg_ridge_mean))
print('\nAccuracy SD For L2 Penalty(penalty of {}): {}'.format(optimal_c_value_ridge, E6_acc_logreg_ridge_sd))

print('\nAUROC Mean For L2 Penalty(penalty of {}): {}'.format(optimal_c_value_ridge, E6_auroc_logreg_ridge_mean))
print('\nAUROC SD For L2 Penalty(penalty of {}): {}'.format(optimal_c_value_ridge, E6_auroc_logreg_ridge_sd))
```

```python
start_rf = time.time()

kf = RepeatedKFold(n_splits=10, n_repeats=10)

test_tar_9 = []

test_acc_rf = []
test_pro_rf = []

for train_index, test_index in kf.split(X=data_bc[predictors],y=data_bc['ReportedConfusion']):

    training_predictors = data_bc[predictors].iloc[train_index]
    training_targets = data_bc['ReportedConfusion'].iloc[train_index]

    test_predictors = data_bc[predictors].iloc[test_index]
    test_targets = data_bc['ReportedConfusion'].iloc[test_index]

    model_rf = RandomForestClassifier()

    test_tar_9.append(test_targets)

    model_rf.fit(training_predictors, training_targets)    
    test_pro_rf.append(model_rf.predict_proba(test_predictors)[:,1])   
    test_acc_rf.append(model_rf.score(test_predictors,test_targets))


E9_mean_test_rf = np.mean(test_acc_rf)
E9_sd_test_rf = statistics.stdev(test_acc_rf)

end_rf = time.time()
elapsedtime_rf= round(end_rf - start_rf, 3)

print('Random Forest Tree-based Accuracy of k-means cross validation with {} folds on test data:'.format(10))
print('\tmean = {}'.format(E9_mean_test_rf))
print('\tstandard deviation = {}\n'.format(E9_sd_test_rf))
print('\tElapsed Time = {} seconds\n'.format(elapsedtime_rf))

E9_y_rf = np.array(test_tar_9).ravel()
E9_score_rf = np.array(test_pro_rf).ravel()
base_fpr = np.linspace(0, 1, 100)

E9_fpr_rf, E9_tpr_rf, E9_thresholds_rf = roc_curve(E9_y_rf, E9_score_rf)

E9_roc_auc_rf, fpr_arr_rf, tpr_arr_rf, tprs_rf = roc_auc(E9_y_rf, E9_score_rf, 9)
E9_mean_roc_auc_rf = np.mean(E9_roc_auc_rf)
E9_sd_roc_auc_rf = statistics.stdev(E9_roc_auc_rf)

tprs_rf = np.array(tprs_rf)
mean_tprs_rf = tprs_rf.mean(axis=0)
std_rf = tprs_rf.std(axis=0)
tprs_upper_rf = np.minimum(mean_tprs_rf + std_rf, 1)
tprs_lower_rf = mean_tprs_rf - std_rf

plt.figure(figsize=(10, 10))
lw = 2
plt.plot(base_fpr, mean_tprs_rf, 'red', label='mean ROC curve (area = %0.2f)' % E9_mean_roc_auc_rf)
plt.fill_between(base_fpr, tprs_lower_rf, tprs_upper_rf, color='grey', alpha=0.3)

for i in range(len(fpr_arr_rf)):    
    plt.plot(fpr_arr_rf[i], tpr_arr_rf[i], color='darkorange')
    
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Random Forest ROC curve for logistic regression')
plt.legend(loc="lower right")
plt.show()
```
