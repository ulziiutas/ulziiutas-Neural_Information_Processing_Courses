```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from matplotlib.pyplot import figure

iris = np.genfromtxt("trace1.txt",delimiter=None) 
print("Total number of voltages:", len(iris))
m=list(map(float,iris))
#plot for spikes
f = []
for i in range(len(m)):
   f.append(i*40)

print("PLOT FOR original")
figure(num=None, figsize=(80, 30), dpi=80, facecolor='w', edgecolor='k')
l0,=plt.plot(f, m, label='plot for spikes')
plt.show()

#threshold with 50

threshold = []
f = []
threshold.append(m[0])
f.append(40)
for i in range(len(m)):
   if i+1 < len(m):
        if m[i] > m[i+1] and m[i+1] < 50:
            threshold.append(m[i+1])
            f.append(len(threshold)*40)
        if m[i] < m[i+1]:
            threshold.append(m[i+1])
            f.append(len(threshold)*40)            
print("PLOT FOR threshold with 50")
figure(num=None, figsize=(80, 30), dpi=80, facecolor='w', edgecolor='k')
l0,=plt.plot(f, threshold, label='plot for spikes')
plt.show()

#threshold with 30
threshold = []
f = []
threshold.append(m[0])
f.append(40)
for i in range(len(m)):
   if i+1 < len(m):
        if m[i] > m[i+1] and m[i+1] < 30:
            threshold.append(m[i+1])
            f.append(len(threshold)*40)
        if m[i] < m[i+1]:
            threshold.append(m[i+1])
            f.append(len(threshold)*40)   
            
print("PLOT FOR threshold with 30")
figure(num=None, figsize=(80, 30), dpi=80, facecolor='w', edgecolor='k')
l0,=plt.plot(f, threshold, label='plot for spikes')
plt.show()

#threshold with 50 and time from 80000 to 90000
trimmedM = m[80000:90000]
threshold = []
f = []
threshold.append(trimmedM[0])
f.append(40)
for i in range(len(trimmedM)):
   if i+1 < len(trimmedM):
        if trimmedM[i] > trimmedM[i+1] and trimmedM[i+1] < 50:
            threshold.append(trimmedM[i+1])
            f.append(len(threshold)*40)
        if trimmedM[i] < trimmedM[i+1]:
            threshold.append(trimmedM[i+1])
            f.append(len(threshold)*40)
            
print("PLOT FOR threshold with 50 and time from 80000 to 90000")
figure(num=None, figsize=(80, 30), dpi=80, facecolor='w', edgecolor='k')
l0,=plt.plot(f, threshold, label='plot for spikes')
plt.show()

#threshold with 30 and time from 80000 to 90000
trimmedM = m[80000:90000]
threshold = []
f = []
threshold.append(trimmedM[0])
f.append(40)
for i in range(len(trimmedM)):
   if i+1 < len(trimmedM):
        if trimmedM[i] > trimmedM[i+1] and trimmedM[i+1] < 30:
            threshold.append(trimmedM[i+1])
            f.append(len(threshold)*40)
        if trimmedM[i] < trimmedM[i+1]:
            threshold.append(trimmedM[i+1])
            f.append(len(threshold)*40)
            
print("PLOT FOR threshold with 30 and time from 80000 to 90000")
figure(num=None, figsize=(80, 30), dpi=80, facecolor='w', edgecolor='k')
l0,=plt.plot(f, threshold, label='plot for spikes')
plt.show()

#without threshold and time from 80000 to 90000
trimmedM = m[80000:90000]
threshold = []
f = []
for i in range(len(trimmedM)):
    threshold.append(trimmedM[i])
    f.append(len(threshold)*40)
            
print("PLOT FOR without threshold and time from 80000 to 90000")
figure(num=None, figsize=(80, 30), dpi=80, facecolor='w', edgecolor='k')
l0,=plt.plot(f, threshold, label='plot for spikes')
plt.show()

# Convert list to numpy array
samples=np.array(m)
mean=np.mean(samples)
var=np.var(samples)
std=np.sqrt(var)
print("This is the standard mean:", mean)
print("This is the standard var:", var)
print("This is the standard deviation:", std)

x=np.linspace(min(samples), max(samples),100)

y_pdf=stats.norm.pdf(x,mean,std)

l1,=plt.plot(x,y_pdf, label='PDF')
plt.show()
```
