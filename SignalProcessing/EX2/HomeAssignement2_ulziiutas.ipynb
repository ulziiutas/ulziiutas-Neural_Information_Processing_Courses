## Part 1: Processing and spectral analysis of Local Field Potentials
Local Field Potentials (LFPs) designate the "low" frequency part of the electrical potential in the extracellular medium of the brain, typically below 200Hz. It reflects many network processes, and is assumed to be generated, to a large extended, by post-synaptic currents entering the neurons.

Let us load some exemplary data recoded from the macaque hippocampus that you can dowload from https://owncloud.tuebingen.mpg.de/index.php/s/REc82bzpCetG27T (update the dataPath variable below before loading).

**Question 2:**
 - (a) Explain what does the "unbiased" option implemented by the acovf and ccovf routines.

If we have $X$ random variables and we want to estimate $E(x_n x_{n-d})$ where $n \in {0...N-1}$, we can average for each data: 

$E(x_n x_{n-d}) = \frac{1}{N-d} \sum^{N-1}_{n=d}x_n x_{n-d}$

This is unbiased estimate and it will converge to true value as you use more data for large $N$. However, when $d$ is closer to the N, estimation will be less acurate as there is not enough data.

As for Biased estimate:

$E(x_n x_{n-d}) = \frac{1}{N} \sum^{N-1}_{n=d}x_n x_{n-d}$

This is biased estimate and it will not converge to true value because it is always dividing by $1/N$ even though there is less number of values.

 - (c) How can we use the acovf and ccovf routines to compute the auto- and cross-correlations to get the functions on an axis spanning both positive and negative lags? Write the code to do that and justify with symmetries shown in the lecture.

Since acovf use np.correlate(a, b, 'full')[n-1:] and cut half of the correlation using [n-1:], we will have a correlation starting from 0 to N-1. To get the negative lag -(N - 1), we can just mirror the positive half.

However, we would get only positve time lag from ccovf(x, y) function. To get the negative time lag:

$\gamma_{XY}(-\tau) = E(X[0]Y[-\tau]) = E(X[\tau]Y[0]) = E(Y[0]X[\tau]) => \gamma_{YX}(\tau) $

 from here we can see that $\gamma_{YX}(\tau)$ with positive $\tau$ is actually equal to the $\gamma_{XY}(-\tau)$ with negative time lag.

ccovf(y, x) also use np.correlate(a, b, 'full')[n-1:] and to get the other half we can get the ccovf(x, y) and flip to get the negative lag of the ccovf.

```python
def acov(x, unbiased_val):
    acovf_pos = acovf(x, unbiased = unbiased_val);
    acovf_full = np.concatenate((np.flip(acovf_pos[1:len(x)]), acovf_pos), axis=None);
    return acovf_full;

def ccov(x, y, unbiased_val):
    ccovf_pos = ccovf(y, x, unbiased = unbiased_val);
    
    ccovf_neg = ccovf(x, y, unbiased = unbiased_val);
    ccovf_neg = np.flip(ccovf_neg[1:len(ccovf_neg)]);
    
    ccovf_full = np.concatenate((ccovf_neg, ccovf_pos), axis=None);
    return ccovf_full;
```

## Part 3: A LTI model of neurovascular coupling

Functional Magnetic Resonance Imaging measures the Blood Oxygen Level Dependent (BOLD) signal related to the concentration of oxyhemoglobin in the brain tissue and is used to assess functional brain activity. 

While the link between neural activity and BOLD is complex, involving several non-linear phenomena, and still elusive, the fMRI community has since long used a linear time invariant system to model this relation. The impulse response of this model is called the Hemodynamic Response Function (HRF), and can be used to investiagte properties of neural activity based on fMRI measurements.

### Reference HRF
We first define a reference HRF in continuous time, similar to those used in popular software such as SPM (https://www.fil.ion.ucl.ac.uk/spm/).

```python
import scipy.stats as stats

def hrf(t):
    pars = [2.5, 2.02, 14.5, 0.247]
    resp = pars[2]*stats.gamma.pdf(t,pars[0],scale = pars[1])-pars[3]*np.exp(-(t-25/2)**2/(12/1.8)**2)
    resp[t<0] = 0
    return resp

t= np.arange(-10,25,.01)

plt.plot(t,hrf(t))
plt.title('HRF model')
plt.xlabel('time (s)')
plt.show()
```

Quantify how varies the mean square error of the estimate of $h$ when we estimate auto- and cross-correlation functions only from on initial segment of data of $30$, $100$, and $200s$. Compare the performance of using biased ("unbiased=False") and unbiased estimates in these cases.

```python
segment_data = np.array([30, 100, 200, 300, 400, 500, 600]);
nrmse = np.zeros((len(segment_data), 2));

plt.figure(figsize=(20, 35));

for i in range(len(segment_data)):
    
    t = np.arange(0, segment_data[i], 1/132);
    y = np.convolve(hrf_sample, x[0:len(t)])[0:len(t)];
    gaussian_noise = np.random.normal(mu, sigma, len(y));
    y = y + gaussian_noise;
    
    acov_x = acov(x[0:len(t)], False);
    ccov_xy = ccov(x[0:len(t)], y, False);
    
    start_ind = int((len(acov_x) - 1)/2 + 1);
    end_ind = int((len(acov_x) - 1)/2 + 1 + p);       
    A = toeplitz(acov_x[start_ind:end_ind]);
    
    start_ind = int((len(ccov_xy) - 1)/2 + 1);
    end_ind = int((len(ccov_xy) - 1)/2 + 1 + p);
    d = ccov_xy[start_ind:end_ind].reshape(1, len(ccov_xy[start_ind:end_ind]));    
    b = np.matmul(d, np.linalg.inv(A));
    
    nrmse[i][0] = np.sqrt(np.sum((b - hrf_sample)**2)/len(b))/(np.max(hrf_sample) - np.min(hrf_sample));
    
    t_1 = np.arange(0, 25,1/132);

    plt.subplot(len(segment_data), 2, i*2 + 1);
    plt.plot(t_1, b[0], label = 'HRF estimation');
    plt.plot(t_1, hrf_sample , label = 'HRF model');
    plt.title('HRF estimate Biased with Time = {}'.format(segment_data[i]));
    plt.xlabel('time (s)');
    plt.legend()
    
    acov_x = acov(x[0:len(t)], True);
    ccov_xy = ccov(x[0:len(t)], y, True);
    
    start_ind = int((len(acov_x) - 1)/2 + 1);
    end_ind = int((len(acov_x) - 1)/2 + 1 + p);      
    A = toeplitz(acov_x[start_ind:end_ind]);
    
    start_ind = int((len(ccov_xy) - 1)/2 + 1);
    end_ind = int((len(ccov_xy) - 1)/2 + 1 + p);
    d = ccov_xy[start_ind:end_ind].reshape(1, len(ccov_xy[start_ind:end_ind]));    
    b = np.matmul(d, np.linalg.inv(A));

    nrmse[i][1] = np.sqrt(np.sum((b - hrf_sample)**2)/len(b))/(np.max(hrf_sample) - np.min(hrf_sample));
    
    plt.subplot(len(segment_data), 2, i*2 + 2);    
    plt.plot(t_1, b[0], label = 'HRF estimation');
    plt.plot(t_1, hrf_sample , label = 'HRF model');
    plt.title('HRF estimate Unbiased with Time = {}'.format(segment_data[i]));
    plt.xlabel('time (s)');
    plt.legend()

plt.show()    
```

```python
plt.figure(figsize=(20, 10));
plt.plot(segment_data[0:2], np.transpose(nrmse)[0][0:2], label='HRF Biased estimation');
plt.plot(segment_data[0:2], np.transpose(nrmse)[1][0:2], label='HRF Unbiased estimation');
plt.title('NMRSE of Unbiased and Biased estimation'.format(segment_data[i]));
plt.xlabel('time (s)');
plt.legend()
```
