```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import statistics
from sklearn.decomposition import PCA
from numba import jit, cuda
from sklearn.decomposition import FactorAnalysis
from scipy.stats import pearsonr

import time
```

```python
n = 200;
mu = 0;
sigma = 1;
z = np.random.normal(mu, sigma, [3, n]);
```

```python
plt.figure(figsize=(20, 5));
plt.suptitle('Histogram of z',fontsize=20);
plt.subplot(1,3,1);
plt.hist(z[0], bins = 20);
plt.subplot(1,3,2);
plt.hist(z[1], bins = 20);
plt.subplot(1,3,3);
plt.hist(z[2], bins = 20);
```

```python
xn1 = z[0];
xn2 = z[1] + 0.001*z[0];
xn3 = 10*z[2];

x = np.array([xn1, xn2, xn3]);
```

```python
plt.figure(figsize=(20, 5))
plt.suptitle('Original Plot',fontsize=20)

plt.subplot(1,3,1)
plt.scatter(x[0], x[1]) 
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.subplot(1,3,2)
plt.scatter(x[0], x[2]) 
plt.xlabel('Feature 1')
plt.ylabel('Feature 3')

plt.subplot(1,3,3)
plt.scatter(x[1], x[2]) 
plt.xlabel('Feature 2')
plt.ylabel('Feature 3')
```

Here, we can see that Feature 1 and Feature 2 has similar value range as can be seen from the equation $x_1 = z_1 and x_2 = z_2 + 0.001z_1$. Also, Feature 3 has very high value range which is -20:30.

```python
pca = PCA(n_components=3);
pca.fit(np.transpose(x));
pca_x = np.transpose(pca.transform(np.transpose(x)));
```

```python
plt.figure(figsize=(20, 5))
plt.suptitle('PCA plots',fontsize=20)

plt.subplot(1,3,1)
plt.scatter(pca_x[0], pca_x[1]) 
plt.xlabel('Principal component 1')
plt.ylabel('Principal component 2')

plt.subplot(1,3,2)
plt.scatter(pca_x[0], pca_x[2]) 
plt.xlabel('Principal component 1')
plt.ylabel('Principal component 3')

plt.subplot(1,3,3)
plt.scatter(pca_x[1], pca_x[2]) 
plt.xlabel('Principal component 2')
plt.ylabel('Principal component 3')
```

<p style="color:red">PCA directions can be better visualized by drawing an error in original plot</p>

```python
pca.explained_variance_
```

Here, We can see that the highest variance principal component has range from -20:30 which is the Feature 3 in our original data points. It is reasonable because it has the highest value range so that it will have highest variance.

```python
r_01 = np.corrcoef(x[0], x[1]);
r_02 = np.corrcoef(x[0], x[2]);
r_12 = np.corrcoef(x[1], x[2]);
```

```python
print('Pearson Correlation xn1-xn2: ', r_01[0][1]);
print('Pearson Correlation xn1-xn3: ', r_02[0][1]);
print('Pearson Correlation xn2-xn3: ', r_12[0][1]);
```

```python
fa = FactorAnalysis(n_components=3)
x_transformed = np.transpose(fa.fit_transform(np.transpose(x)))
```

```python
plt.figure(figsize=(20, 5));
plt.suptitle('FA plots',fontsize=20);

plt.subplot(1,3,1);
plt.scatter(x_transformed[0], x_transformed[1]); 
plt.xlabel('FA component 1');
plt.ylabel('FA component 2');

plt.subplot(1,3,2);
plt.scatter(x_transformed[0], x_transformed[2]); 
plt.xlabel('FA component 1');
plt.ylabel('FA component 3');

plt.subplot(1,3,3);
plt.scatter(x_transformed[1], x_transformed[2]);
plt.xlabel('FA component 2');
plt.ylabel('FA component 3');
```
