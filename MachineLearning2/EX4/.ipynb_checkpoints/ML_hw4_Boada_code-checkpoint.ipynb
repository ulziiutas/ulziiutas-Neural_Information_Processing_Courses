```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
N = 20 # number datapoints
```

```python
# import datapoints
x_train = np.array(open('Datafiles-Exer-4/Traindatax.txt', 'r').read().split()).astype(float)
y_train = np.array(open('Datafiles-Exer-4/Traindatay.txt', 'r').read().split()).astype(float)
x_test = np.array(open('Datafiles-Exer-4/Testdatax.txt', 'r').read().split()).astype(float)
y_test = np.array(open('Datafiles-Exer-4/Testdatay.txt', 'r').read().split()).astype(float)

# visualization
plt.subplots(figsize=(10,3))
plt.subplot(121)
plt.scatter(x_train, y_train)
plt.title('Train data')
plt.subplot(122)
plt.scatter(x_test, y_test)
plt.title('Test data')
plt.show()
```

### Maximum likelihood solution
Implement derived solution from 1a:

$\mathbf { w } _ { \mathrm { ML } } = \left( \mathbf { \Phi } ^ { \mathrm { T } } \mathbf { \Phi } \right) ^ { - 1 } \mathbf { \Phi } ^ { \mathrm { T } } \mathbf { t }$

where:

$\mathbf { \Phi } = \left( \begin{array} { c c c c } { \phi _ { 0 } \left( \mathbf { x } _ { 1 } \right) } & { \phi _ { 1 } \left( \mathbf { x } _ { 1 } \right) } & { \cdots } & { \phi _ { M - 1 } \left( \mathbf { x } _ { 1 } \right) } \\ { \phi _ { 0 } \left( \mathbf { x } _ { 2 } \right) } & { \phi _ { 1 } \left( \mathbf { x } _ { 2 } \right) } & { \cdots } & { \phi _ { M - 1 } \left( \mathbf { x } _ { 2 } \right) } \\ { \vdots } & { \vdots } & { \ddots } & { \vdots } \\ { \phi _ { 0 } \left( \mathbf { x } _ { N } \right) } & { \phi _ { 1 } \left( \mathbf { x } _ { N } \right) } & { \cdots } & { \phi _ { M - 1 } \left( \mathbf { x } _ { N } \right) } \end{array} \right)$

$\phi _ { 0 } (  x  ) = 1$

$\phi _ { m } (  x  ) = \exp \left( - \left|  x  -  \mu  _ { m } \right| ^ { 2 } / \left( 2 \sigma _ { \phi } ^ { 2 } \right) \right) $

```python
beta = 50 # noise precision
sigma = [1, 10, .1] # variance from Gaussian RBF 
```

```python
# Define de Gaussian RBFs
M = 9 # number of RGBs (arbitrary)
mu = np.linspace(0,6.5,M) # Gaussian RBFs centers
def phi(m, x, var=sigma[0], centers=mu):
    if m==0:
        return np.ones(np.size(x))
    else:
        return np.exp(-(x-centers[m])**2 / (2*var**2))

# plot all basis functions (uniformly separated and equally weighted)
plt.subplots(figsize=(15,4))
x_array = np.arange(0,8,.01)
for s in range(3):
    plt.subplot(1,3,s+1)
    for m in range(M):
        plt.plot(x_array, phi(m, x=x_array, var=sigma[s]))
        plt.title('$\sigma$ = {}'.format(sigma[s]))
print('Basis functions (equally weighted):')
plt.show()
```

```python
# Build the Gramian matrices
gram = np.ones((N,M,3))
for s in range(3):
    for n, x in enumerate(x_train):
        for m in range(M):
            gram[n,m,s] = phi(m, x, var=sigma[s])

pd.DataFrame(gram[:,:,0]).head()
```

```python
# ML solution for the weight vector
w_ml = np.zeros((M,3))
for s in range(3):
    w_ml[:,s] = np.linalg.inv(gram[:,:,s].T@gram[:,:,s])@gram[:,:,s].T@y_train
    print('Weights vector for sigma={}, w:\n{}\n'.format(sigma[s], w_ml[:,s]))

# plot all basis function and their sum, i.e. the regression
plt.subplots(figsize=(15,4))
for s in range(3):
    plt.subplot(1,3,s+1)
    y_sum = np.zeros(len(x_array))
    for m in range(M):
        plt.plot(x_array, w_ml[m,s]*phi(m, x=x_array, var=sigma[s]))
        y_sum += w_ml[m,s]*phi(m, x=x_array,var=sigma[s])

    plt.plot(x_array, y_sum, 'k', linewidth=5, label='Regression')
    plt.scatter(x_train,y_train)
#     plt.ylim((-5,5))
    plt.title('$\sigma$ = {}'.format(sigma[s]))
    plt.legend()
    
print('Basis functions (optimally weighted):')
plt.show()
```

```python
# Prediction function
def pred(x, w_ml=w_ml[:,0], gram=gram[:,:,0]):
    N = np.size(x)
    M = np.size(w_ml)
    y_pred = np.zeros(N)
    
    # Build Gramian matrix
    gram = np.ones((N,M))
    for n, x_n in enumerate(x):
        for m in range(M):
            gram[n,m] = phi(m, x_n)
            
    # Predict values        
    y_pred = w_ml.T@gram.T
    return y_pred

# Error calculator (sum of absolute residuals)
def error(y_real, y_pred):
    return np.abs(y_real-y_pred).sum()
```

```python
# Apply the functions to predict the training and testing datapoints for the 3 width parameters

y_train_pred = np.zeros((len(x_train), 3))
y_test_pred = np.zeros((len(x_test), 3))
E_train = np.zeros(3)
E_test = np.zeros(3)

for s in range(3):
    
    # Predict training datapoints
    y_train_pred[:,s] = pred(x_train, w_ml=w_ml[:,s], gram=gram[:,:,s])
    E_train[s] = error(y_train, y_train_pred[:,s])
    # Predict testing datapoints
    y_test_pred[:,s] = pred(x_test, w_ml=w_ml[:,s], gram=gram[:,:,s])
    E_test[s] = error(y_test, y_test_pred[:,s])

    # Visualization
    plt.subplots(figsize=(10,3))

    plt.subplot(121)
    plt.scatter(x_train, y_train, label='true')
    plt.scatter(x_train, y_train_pred[:,s], c='orange', label='predicted')
    plt.title('Training datapoints prediction, $\sigma$ = {}'.format(sigma[s]))
    plt.legend()

    plt.subplot(122)
    plt.scatter(x_test, y_test, label='true')
    plt.scatter(x_test, y_test_pred[:,s], c='orange', label='predicted')
    plt.title('Testing datapoints prediction, $\sigma$ = {}'.format(sigma[s]))
    plt.legend()

    plt.show()

    print('Errors (sum of absolute residuals):\n\
    E_train = {}\n\
    E_test = {}\n---------------------------------------------------------'
          .format(np.round(E_train[s],2), np.round(E_test[s],2)))
```

As we can see from the plots an errors, the width of the Gaussian RBFs is critical to get a good prediction. Im our case, $\sigma = 1$ gives the best result because that specific width allows the basis functions to fit the somehow sinusoidal shape of the datapoints.

### Bayesian regression solution

Now, our weigths vector is distributed according to the next distribution

$p ( \mathbf { w } | \mathbf { t } ) = p \left( \mathbf { t } _ { \mathbf { N } } \mathbf { w } \right) p ( \mathbf { w } ) = \mathcal { N } \left( \mathbf { w } | \mathbf { m } _ { N } , \mathbf { S } _ { N } \right)$

where:

$\begin{aligned} \mathbf { m } _ { N } & = \beta \mathbf { S } _ { N } \mathbf { \Phi } ^ { \mathrm { T } } \mathbf { t } \\ \mathbf { S } _ { N } ^ { - 1 } & = \alpha \mathbf { I } + \beta \mathbf { \Phi } ^ { \mathrm { T } } \boldsymbol { \Phi } \end{aligned}$

```python
alpha = [1, 1e-8] # prior precision
```

In order to keep it simple and allow an easier comparison, from now on I fix the RBFs width to $\sigma = 1$ and play with the prior precision, $\alpha$.

```python
w_bayes = np.zeros((M,2))
for a in range(2):
    # Calculate the parameters of our the weight vector distribution
    S_N = np.linalg.inv(alpha[a]*np.identity(M) + beta*gram[:,:,0].T@gram[:,:,0])
    m_N = beta*S_N@gram[:,:,0].T@y_train

    # Drawing the most likely weight vector, which actually is w_bayes = m_N
    w_bayes[:,a] = np.random.multivariate_normal(m_N, np.zeros((M,M)))

    print('Weights vector for sigma={} and alpha={}:\n{}\n'.format(sigma[0], alpha[a], w_bayes[:,a]))
```

```python
# plot all basis function and their sum, i.e. the regression
plt.subplots(figsize=(15,4))
for a in range(2):
    y_sum = np.zeros(len(x_array))
    for m in range(M):
        plt.subplot(1,2,a+1)
        plt.plot(x_array, w_bayes[m,a]*phi(m, x=x_array))
        y_sum += w_bayes[m,a]*phi(m, x=x_array)
    plt.plot(x_array, y_sum, 'k', linewidth=5, label='Regression')
    plt.scatter(x_train,y_train)
    plt.ylim((-1.5,3.2))
    plt.title('$\\alpha$  = {}'.format(alpha[a]))
    plt.legend()
    

print('Basis functions (optimally weighted):')
plt.show()
```

Given 2 different priors, the obtained weights vector are quite different, but the regression ends up being similar.

```python
# Apply the functions to predict the training and testing datapoints for the 3 width parameters
y_train_pred = np.zeros((len(x_train), 2))
y_test_pred = np.zeros((len(x_test), 2))
E_train = np.zeros(2)
E_test = np.zeros(2)

for a in range(2):
    
    # Predict training datapoints
    y_train_pred[:,a] = pred(x_train, w_ml=w_bayes[:,a], gram=gram[:,:,0])
    E_train[a] = error(y_train, y_train_pred[:,a])
    # Predict testing datapoints
    y_test_pred[:,a] = pred(x_test, w_ml=w_bayes[:,a], gram=gram[:,:,0])
    E_test[a] = error(y_test, y_test_pred[:,a])

    # Visualization
    plt.subplots(figsize=(10,3))

    plt.subplot(121)
    plt.scatter(x_train, y_train, label='true')
    plt.scatter(x_train, y_train_pred[:,a], c='orange', label='predicted')
    plt.title('Training datapoints prediction, $\\alpha$ = {}'.format(alpha[a]))
    plt.legend()

    plt.subplot(122)
    plt.scatter(x_test, y_test, label='true')
    plt.scatter(x_test, y_test_pred[:,a], c='orange', label='predicted')
    plt.title('Testing datapoints prediction, $\\alpha$ = {}'.format(alpha[a]))
    plt.legend()

    plt.show()

    print('Errors (sum of absolute residuals):\n\
    E_train = {}\n\
    E_test = {}\n---------------------------------------------------------'
          .format(np.round(E_train[a],2), np.round(E_test[a],2)))
```

Although the priors were different, N=20 datapoints it's enough to get so good weights vector that the regression ends up being really similar. The errors similarity confirms this statement.
