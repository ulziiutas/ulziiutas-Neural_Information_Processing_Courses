```python
import numpy as np
import mpmath as mpmath
import math as math
import matplotlib.pyplot as plt
import seaborn as sns   # not used here but in general a useful library for plotting
%matplotlib inline

d = 2 * 10**(-6)
l = 100 * 10**(-6)
em = 0
rm = 1
ra = 1
cm = 10**(-2)
deltaT = 0.1 * 10**(-3)
ie = -50

R = rm /(math.pi * d * l)
C = cm * math.pi * d * l
```

# <h2>Exercise 3.1</h2>
Using backward Euler method we get the following Equation. 

```python
V = []
t = []

for i in range(1000):
    mult =  float("%0.5f" % (i * deltaT))
    if mult < deltaT:
        V.append(float("%0.5f" % (ie * deltaT / C)) / (deltaT/(R * C) + 1))
        t.append(mult)
    else:        
        V.append((float("%0.5f" % (ie * deltaT / C)) + V[i-1]) / (deltaT/ (R * C) + 1))
        t.append(mult)
```

```python
plt.figure(figsize=(5, 5))
plt.plot(t, V, label='time and voltage relationship plot')
plt.title(r"Exercise 3.1. The time course for V (t) for a step input Ie(t)""\n")
plt.xlabel(r'time [s]')
plt.ylabel(r'Voltage [V]')
plt.legend()
plt.show()
```

As the signal propagation depends on time constant and length constant. Here, time constant is dependent on cm and rm.
As we have increased the resistance rm, we get faster signal propagation thus we have steep curve in green. But when we decrease the capacitance cm we get less steep curve than the line blue which is our original curve. Because now time constant gets smaller and we have slower signal propagation.
