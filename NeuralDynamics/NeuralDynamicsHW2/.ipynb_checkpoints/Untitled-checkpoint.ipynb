```python
import numpy as np
import mpmath as mpmath
import math as math
import matplotlib.pyplot as plt
import seaborn as sns   # not used here but in general a useful library for plotting
%matplotlib inline

d = 2 * 10**(-6)
l = 100 * 10**(-6)
em = 0
rm = 1
ra = 1
cm = 10**(-2)
deltaT = 0.1 * 10**(-3)
ie = -50

R = (2/math.pi) * d**(-1.5) * mpmath.coth(l)
C = cm * math.pi * (d**2) / 4

print(R)
print(C)
```

```python
V = []
t = []

for i in range(200):
    mult =  float("%0.5f" % (i * deltaT))
    if mult < deltaT:
        V.append(float("%0.5f" % (ie * deltaT)) / (deltaT/R + C))
        t.append(mult)
    else:        
        V.append((float("%0.5f" % (ie * deltaT)) + C * V[i-1]) / (deltaT/R + C))
        t.append(mult)
print(V)
```

```python
plt.figure(figsize=(15, 15))
plt.plot(t, V, label='time and voltage relationship plot')
plt.title(r"              The time course for V (t) for a step input Ie(t)")
plt.xlabel(r'V [V]')
plt.ylabel('t [s]')
plt.ylim(0, max(t))
print(max(t))
plt.legend()
plt.show()
```
