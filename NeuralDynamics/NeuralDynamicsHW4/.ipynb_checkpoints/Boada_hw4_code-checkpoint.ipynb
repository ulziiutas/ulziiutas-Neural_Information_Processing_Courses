### Exercise 1. Simulation of multi compartment model of passive neurite.

```python
# PARAMETERS
Cm = 62.8e-12 # Membrane capacitance [F]
Rm = 1.59e9 # Membrance resistance [Ohm]
Ra = 0.0318e9 # Axial resistance [Ohm]
E_L= 0 # Membrane potential at rest [V]
```

#### 1.1

Derived DEQ for the potential at compartment j in a multi compartment model with an arbitrary number of compartments:

General case:

$\frac{dV_j}{dt}=\frac{1}{C_m}(I_e-V_j(\frac{2}{R_a}+\frac{1}{R_m})+\frac{V_{j-1}}{R_a}+\frac{V_{j+1}}{R_a}+\frac{E_L}{R_m})$

Sealed compartment (For the first compartment, $j=1$)

$\frac{dV_j}{dt}=\frac{1}{C_m}(I_e-V_j(\frac{1}{R_a}+\frac{1}{R_m})+\frac{V_{j-1}}{R_a}+\frac{V_{j+1}}{R_a}+\frac{E_L}{R_m})$

Killed compartment (For the last compartment, $j=N$)

$\frac{dV_j}{dt}=\frac{1}{C_m}(I_e-V_j(\frac{2}{R_a}+\frac{1}{R_m})+\frac{V_{j-1}}{R_a}+\frac{V_{j+1}}{R_a}+\frac{E_L}{R_m})$

$I_e$ is 0 everywhere exept from the compartment where the current is injected

```python
# PARAMETERS
t = 1 # desired simulation time (arbitrary) [s]
dt = 0.1e-3 # Time step (arbitrary) [s] 
n = np.int(t/dt) # number of samples (dt seconds/sample * n samples = t second)
t_array=np.arange(0,t,dt) # time array, useful for the plot.
j_array=np.arange(0,N) # compartments array, useful for the plot.
N = 50 # Number of compartments

# Current step function parameters
I0 = 10e-12 # Input current [A]
je = 20 # Compartment at which current is injected
te = 20e-3 # Time at which current injection starts [s]
# Current step function
def Ie(t, j=je, te=te): 
    if t < te or j != je:
        return 0
    if t >= te and j == je:
        return I0 
```

```python
# SIMULATION

# Initialize matrix V to save the voltages of all compartment over time
V = np.zeros((N,n))

# Build matrix
## Initialize a matrix of zeros with N x N size
M = np.zeros((N,N))
## Fill the main diagonal with the general term
np.fill_diagonal(M, -((2/Ra)+(1/Rm)))
## Correct for the first term (sealed) and last term (killed) of the diagonal
M[0,0] = -((1/Ra)+(1/Rm))
M[N-1,N-1] = 0 
## Fill the offset diagonals: 1 above and 1 underneath
np.fill_diagonal(M[1:,:], 1/Ra) # M[1:] skips the 1st row. Main diagonal "becomes" the -1 offset diagonal
np.fill_diagonal(M[:,1:], 1/Ra) # M[:,1:] skips the 1st column. Main diagonal "becomes" the +1 offset diagonal

# print(M)  # We can check how the matrix has been correctly built

# Computation

for i in range(n):
    ti = float("%0.7f" % (i*dt)) # time step index, it moves 1 step for each iteration
    for j in range (N-1):
        if i==0:  # for the 1st iteration
            V[j][i] = 0 + (dt/Cm) * (Ie(ti) - 0*((2/Ra)+(1/Rm)) + (0/Ra) + (0/Ra) + (E_L/Rm))
        else:
            if j == 0:
                V[j][i] = V[j][i-1] + (dt/Cm) * (Ie(ti) - V[j][i-1]*((1/Ra)+(1/Rm)) + 0 + (V[j+1][i-1]/Ra) + (E_L/Rm)) 
            if j == N:
                V[j][i] = 0
            else:
                V[j][i] = V[j][i-1] + (dt/Cm) * (Ie(ti) - V[j][i-1]*((2/Ra)+(1/Rm)) + (V[j-1][i-1]/Ra) + (V[j+1][i-1]/Ra) + (E_L/Rm)) 
    
```

```python
# 2D VISUALIZATION

plt.figure(figsize=(15, 6))
# plt.suptitle('Voltage output signals', fontsize=16)
# for k in range(6):
#     plt.subplot(2,3,k+1)
#     plt.plot(t,V_f[k])
#     plt.title('f={}Hz'.format(f[k]))
#     plt.xlabel('t [s]', fontsize=12)
#     plt.ylabel('V [V]', fontsize=12)
# plt.subplots_adjust(top=0.89, bottom=0.08, left=0.05, right=0.95, hspace=0.40, wspace=0.35)
# plt.show()   

# Plot all of the compartments response:
plt.subplot(1,2,1)
for j in range(N):
    plt.plot(t_array,V[j], label='Compartment {}'.format(j+1))
# plt.legend()
plt.show()

plt.figure(figsize=(15, 6))

# Plot for significant compartments
plt.subplot(1,2,2)
# Compartment 1, first and sealed
plt.plot(t_array,V[0], label='Compartment {}'.format(1))
# Compartment 18, before and near where current is injected
plt.plot(t_array,V[je-10], label='Compartment {}'.format(je-10))
# Compartment je, where current is injected
plt.plot(t_array,V[je], label='Compartment {}'.format(je))
# Compartment 22, after and near where current is injected
plt.plot(t_array,V[je+10], label='Compartment {}'.format(je+10))
# Compartment 50, last and killed
plt.plot(t_array,V[N-1], label='Compartment {}'.format(N))
plt.legend()
plt.show()
```

As the current is injected in compartment 20 and we considered a total of 50 compartment, the system is not symmetric. This is why the response at two compartment equally distanced from je, do not show the same response. Moreover, if current was injected in the middle (compartment 25), the response at two equally distanced compartments from je would still not be equal because the terminals do not have the same condition (first=sealed, last=killed)

```python
# 3D VISUALIZATION

# Build matrices to plot
t_matrix = np.zeros((N,n))
for i in range(N):
    t_matrix[i] = t_array
    
j_matrix = np.zeros((n,N))
for i in range(n):
    j_matrix[i] = j_array
j_matrix_transposed = np.transpose(j_matrix)

# check shapes of matrices (must coincide)
print("shape of t_array = {}".format(t_matrix.shape))
print("shape of j_array = {}".format(j_matrix_transposed.shape))
print("shape of V = {}".format(V.shape))

# plot functions
fig = plt.figure(figsize=(7,6))
ax = fig.gca(projection='3d')
fig.suptitle("Title", fontsize=16, y=0.90)
ax.set_xlabel('Time [s]', fontsize=15)
ax.set_ylabel('Compartment', fontsize=15)
ax.set_zlabel('Voltage [V]', fontsize=15)
surf = ax.plot_surface(t_matrix, j_matrix_transposed, V, cmap=cm.coolwarm, linewidth=0, antialiased=False)
plt.show()
```
