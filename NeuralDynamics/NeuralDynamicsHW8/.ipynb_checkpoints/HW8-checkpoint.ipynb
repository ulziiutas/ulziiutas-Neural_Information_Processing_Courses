## E1.4

```python
tau = 10
a = 1.0
b = 0.6
d = 2
k0 = 4

A = 10
B = -10
M = 200

deltaX = (B - A) / M
deltaTime = 0.1

noise = np.random.normal(0,0.01**2,200)

def interaction_kernel(x, a_, k0_):    
    exp = np.exp(-(x**2) / (4 * b**2))
    cos = np.cos(k0 * x)    
    return (a_ * exp * cos) / (np.sqrt(np.pi) * b)

def stimulation(i):    
    exp = np.exp(-(x[i]**2) / (4 * d**2))
    return exp / (2 * np.sqrt(np.pi) * d)

def stimulation_noisy(i):
    exp = np.exp(-(x[i]**2) / (4 * d**2))
    return (exp / (2 * np.sqrt(np.pi) * d) + noise[i])
    

def x_n(n):
    return A + n * deltaX

def riemann_sum(u_x_t, x_index, t_index, a_, k0_):
    r_sum = 0
    for i in range(M):
        r_sum = r_sum + interaction_kernel(x[x_index] - x[i], a_, k0_) * u_x_t[i][t_index] * deltaX
    return r_sum

x = np.zeros(M)

for i in range(M):
    x[i] = x_n(i)
    
    
def u_k(a_, k0_):    
    
    t = np.arange(0, 20, deltaTime)
    
    u_x_t = np.zeros(len(t) * M).reshape(M, len(t))
    
    for i in range(M):
        u_x_t[i][0] = (deltaTime / tau) * (stimulation(i))
    
    for i in range(M):
        
        for j in range(1, len(t)):
            
            u_x_t[i][j] = (deltaTime / tau) * (-u_x_t[i][j - 1] + riemann_sum(u_x_t, i, j - 1, a_, k0_) + stimulation(i)) + u_x_t[i][j - 1]
               
    return u_x_t, x, t
    
def u_k_noisy(a_, k0_):    
    
    t = np.arange(0, 20, deltaTime)
    
    u_x_t = np.zeros(len(t) * M).reshape(M, len(t))
    
    for i in range(M):
        u_x_t[i][0] = (deltaTime / tau) * (stimulation_noisy(i))
    
    for i in range(M):
        
        for j in range(1, len(t)):
            
            u_x_t[i][j] = (deltaTime / tau) * (-u_x_t[i][j - 1] + riemann_sum(u_x_t, i, j - 1, a_, k0_) + stimulation_noisy(i)) + u_x_t[i][j - 1]
               
    return u_x_t, x, t
```

```python
u, x, t = u_k(a, k0)
```

```python
t_x_1 = np.zeros([len(t), len(x)])
for i in range(len(t)):    
    for k in range(len(x)):
        t_x_1[i][k] = t[i]

x_t_1 = np.zeros([len(t), len(x)])
for i in range(len(t)):
    x_t_1[i] = x

fig = plt.figure(figsize=(17,16))
ax = fig.gca(projection='3d')
fig.suptitle("Neural Field agains Time and Spatial plot", fontsize=16, y=0.90)
ax.set_xlabel('Time[ms]', fontsize=15)
ax.set_ylabel('Spatial Distance', fontsize=15)
ax.set_zlabel('Neural Field', fontsize=15)
surf = ax.plot_surface(t_x_1, x_t_1, u.transpose(), cmap=cm.coolwarm, linewidth=0, antialiased=False)
plt.show()
```

The Linear Neural field increases over time but decreases over spatially. And Overtime it starts to saturate.

```python
c = 1
d1 = 0.5
v = 0.1
a = 1.0
k0 = 4.0

def stimulation_2(i, t, v):    
    exp = np.exp(-((x[i] - v * t)**2) / (4 * d1**2))
    return (c * exp) / (2 * np.sqrt(np.pi) * d1)

def interaction_kernel_2(x):    
    exp = np.exp(-c * np.abs(x))
    sign = 0
    if x >= 0:
        sign = 1
    else:
        sign = -1
    return exp * sign

def riemann_sum_2(u_x_t, x_index, t_index):
    r_sum = 0
    for i in range(M):
        r_sum = r_sum + interaction_kernel_2(x[x_index] - x[i]) * u_x_t[i][t_index] * deltaX
    return r_sum

def u_k_s2(v):    
    
    t = np.arange(0, 20, deltaTime)
    
    u_x_t = np.zeros(len(t) * M).reshape(M, len(t))
    
    for i in range(M):
        u_x_t[i][0] = (deltaTime / tau) * (stimulation_2(i, 0, v))
    
    for i in range(M):
        
        for j in range(1, len(t)):
            
            u_x_t[i][j] = (deltaTime / tau) * (-u_x_t[i][j - 1] + riemann_sum_2(u_x_t, i, j - 1) + stimulation_2(i, t[j], v)) + u_x_t[i][j - 1]
               
    return u_x_t, x, t
```

## 2.1

```python
def interaction_kernel_e2(x, a_, b_, A_, B_):    
    
    if np.abs(x) <= a_:
        return A_
    elif a < np.abs(x) <= b_:
        return -B_
    else:
        return 0
    return 0

def stimulation_e2(i, C_, d_):    
    if np.abs(x[i]) <= d_:
        C_ * (1 - np.abs(x[i]) / d_)
    else:
        return 0
    return 0
    
def riemann_sum_e2(u_x_t, x_index, t_index, a_, b_, A_, B_):
    r_sum = 0
    for i in range(M):
        if u_x_t[i][t_index] < 0:            
            r_sum = r_sum + interaction_kernel_e2(x[x_index] - x[i], a_, b_, A_, B_) * 0 * deltaX
        else:
            r_sum = r_sum + interaction_kernel_e2(x[x_index] - x[i], a_, b_, A_, B_) * 1 * deltaX
            
    return r_sum

x = np.zeros(M)

for i in range(M):
    x[i] = x_n(i)
    
    
def u_k_e2(a_, b_, A_, B_, C_, d_, h_, deltaTime_):    
    
    t = np.arange(0, 1, deltaTime_)
    
    u_x_t = np.zeros(len(t) * M).reshape(M, len(t))
    
    for i in range(M):
        u_x_t[i][0] = -h_
    
    for i in range(M):
        
        for j in range(1, len(t)):
            
            u_x_t[i][j] = (deltaTime_ / tau) * (-u_x_t[i][j - 1] + riemann_sum_e2(u_x_t, i, j - 1, a_, b_, A_, B_) + stimulation_e2(i, C_, d_) - h_) + u_x_t[i][j - 1]
               
    return u_x_t, x, t
    
def u_k_e2_1(a_, b_, A_, B_, C_, d_, h_, deltaTime_):    
    
    t = np.arange(0, 1, deltaTime_)
    
    u_x_t = np.zeros(len(t) * M).reshape(M, len(t))
    
    for i in range(M):
        u_x_t[i][0] = 3.3 * stimulation_e2(i, C_, d_) - h_
    
    for i in range(M):
        
        for j in range(1, len(t)):
            
            u_x_t[i][j] = (deltaTime_ / tau) * (-u_x_t[i][j - 1] + riemann_sum_e2(u_x_t, i, j - 1, a_, b_, A_, B_) + stimulation_e2(i, C_, d_) - h_) + u_x_t[i][j - 1]
               
    return u_x_t, x, t
    
def u_k_e2_2(a_, b_, A_, B_, C_, d_, h_, deltaTime_):    
    
    t = np.arange(0, 1, deltaTime_)
    
    u_x_t = np.zeros(len(t) * M).reshape(M, len(t))
    
    for i in range(M):
        u_x_t[i][0] = -h_
    
    for i in range(M):
        
        for j in range(1, len(t)):
            
            u_x_t[i][j] = (deltaTime_ / tau) * (-u_x_t[i][j - 1] + 0.1 * riemann_sum_e2(u_x_t, i, j - 1, a_, b_, A_, B_) + stimulation_e2(i, C_, d_) - h_) + u_x_t[i][j - 1]
               
    return u_x_t, x, t

def u_k_e2_3(a_, b_, A_, B_, C_, d_, h_, deltaTime_):    
    
    t = np.arange(0, 1, deltaTime_)
    
    u_x_t = np.zeros(len(t) * M).reshape(M, len(t))
    
    for i in range(M):
        u_x_t[i][0] = (deltaTime / tau) * (stimulation_e2(i, C_, d_))
    
    for i in range(M):
        
        for j in range(1, len(t)):
            
            u_x_t[i][j] = (deltaTime_ / tau) * (-u_x_t[i][j - 1] + 0.1 * riemann_sum_e2(u_x_t, i, j - 1, a_, b_, A_, B_) + stimulation_e2(i, C_, d_) - h_) + u_x_t[i][j - 1]
               
    return u_x_t, x, t
```
