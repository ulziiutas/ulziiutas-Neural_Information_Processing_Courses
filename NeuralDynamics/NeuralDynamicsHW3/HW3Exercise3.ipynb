```python
import numpy as np
import mpmath as mpmath
import math as math
import matplotlib.pyplot as plt
%matplotlib inline

Em = 0
Rm = 265e6
Ra = 7e6
Cm = 75e-12

deltaT = 0.1e-3
te = 0.4
ts = 0.44

print(Cm,Ra)
```

```python
def ie_t(t, te, ts):
    if t < te:
        return 0
    if te <= t and t < ts:
        return -100e-12
    if ts <= t:
        return 0
```

```python
def ie_t_sinusoid(t, f):
    return math.sin(2 * math.pi * f * t) * 100 * 10**(-12)
```

```python
def v1_t_plus_deltaT(deltaT, te, ts, Cm, Rm, Ra, max_range, IS_ie_t_sinusoid, f):
    V1 = []
    V2 = []
    t = []
    for i in range(max_range):
        mult =  float("%0.7f" % (i * deltaT))
        if mult < deltaT:
            if IS_ie_t_sinusoid == True:    
                V1.append(ie_t_sinusoid(mult, f) * (deltaT/Cm))
                V2.append(0)
                t.append(mult)
            else:           
                V1.append(0)
                V2.append(0)     
                t.append(mult)
        else:
            if IS_ie_t_sinusoid == True:        
                V1.append((ie_t_sinusoid(mult, f) - V1[i-1] * (1/Rm + 1/Ra) + V2[i-1]/Ra) * (deltaT/Cm) + V1[i-1])
                V2.append((V1[i-1]/Ra - V2[i-1] * (1/Rm + 1/Ra)) * (deltaT/Cm) + V2[i-1])
                t.append(mult)    
            else:                
                V1.append((ie_t(mult, te, ts) - V1[i-1] * (1/Rm + 1/Ra) + V2[i-1]/Ra) * (deltaT/Cm) + V1[i-1])
                V2.append((V1[i-1]/Ra - V2[i-1] * (1/Rm + 1/Ra)) * (deltaT/Cm) + V2[i-1])
                t.append(mult)         
    return V1, V2, t
```

```python
V1_ra1, V2_ra1, t = v1_t_plus_deltaT(deltaT, te, ts, Cm, Rm, Ra, 6000, False, 0)
plt.plot(t, V1_ra1, label='V1(t) against t')
plt.plot(t, V2_ra1, label='V2(t) against t')

plt.title(r"Exercise 3.2 V1 (t) and V2 (t) against t with Ra = 7MOhm.""\n")
plt.xlabel(r'time [s]')
plt.ylabel(r'Voltage [V]')
plt.legend()
plt.show()

Ra = 256e6
V1_ra2, V2_ra2, t = v1_t_plus_deltaT(deltaT, te, ts, Cm, Rm, Ra, 6000, False, 0)
plt.plot(t, V1_ra2, label='V1(t) against t')
plt.plot(t, V2_ra2, label='V2(t) against t')

plt.title(r"Exercise 3.2 V1 (t) and V2 (t) against t with Ra = 265MOhm.""\n")
plt.xlabel(r'time [s]')
plt.ylabel(r'Voltage [V]')
plt.legend()
plt.show()

Ra = 30e9
V1_ra3, V2_ra3, t = v1_t_plus_deltaT(deltaT, te, ts, Cm, Rm, Ra, 6000, False, 0)
plt.plot(t, V1_ra3, label='V1(t) against t')
plt.plot(t, V2_ra3, label='V2(t) against t')

plt.title(r"Exercise 3.2 V1 (t) and V2 (t) against t with Ra = 30GOhm.""\n")
plt.xlabel(r'time [s]')
plt.ylabel(r'Voltage [V]')
plt.legend()
plt.show()

plt.plot(t, V1_ra1, label='V1(t) against t with Ra = 7MOhm')
plt.plot(t, V1_ra2, label='V1(t) against t with Ra = 256MOhm')
plt.plot(t, V1_ra3, label='V1(t) against t with Ra = 30GOhm')

plt.title(r"Exercise 3.2 V1 (t) against t with varying Ra""\n")
plt.xlabel(r'time [s]')
plt.ylabel(r'Voltage [V]')
plt.legend()
plt.show()

plt.plot(t, V2_ra1, label='V2(t) against t with Ra = 7MOhm')
plt.plot(t, V2_ra2, label='V2(t) against t with Ra = 256MOhm')
plt.plot(t, V2_ra3, label='V2(t) against t with Ra = 30GOhm')

plt.title(r"Exercise 3.2 V2 (t) against t with varying Ra""\n")
plt.xlabel(r'time [s]')
plt.ylabel(r'Voltage [V]')
plt.legend()
plt.show()
```
